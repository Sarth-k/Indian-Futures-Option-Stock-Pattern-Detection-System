```python


warnings.filterwarnings('ignore')

# Setup logging
logging.basicConfig(
    level=logging.INFO,
    format='%(asctime)s - %(levelname)s - %(message)s',
    handlers=[
        logging.FileHandler('stock_detector.log'),
        logging.StreamHandler()
    ]
)

@dataclass
class TradingSignal:
    symbol: str
    pattern: str
    signal_type: str
    confidence: float
    entry_price: float
    target_price: float
    stop_loss: float
    risk_reward_ratio: float
    volume_confirmation: bool
    sentiment_score: float
    market_regime: str
    timestamp: datetime

class AdvancedStockPatternDetector:
    def __init__(self, config_file='trading_config.json'):
        self.config = self.load_config(config_file)
        self.setup_database()

        # Enhanced Indian stock universe (Top 200 NSE stocks)
        self.nse_stocks = self.load_nse_universe()
        self.bse_stocks = self.load_bse_universe()

        # ML Models
        self.pattern_classifier = None
        self.anomaly_detector = IsolationForest(contamination=0.1, random_state=42)
        self.lstm_model = None

        # Market data
        self.market_data = {}
        self.sentiment_data = {}
        self.option_data = {}

        # Risk management
        self.portfolio_exposure = 0.0
        self.max_portfolio_risk = 0.15  # 15% max portfolio risk

        # Real-time features
        self.active_alerts = {}
        self.paper_trades = []

        self.initialize_models()

    def load_config(self, config_file):
        """Load configuration from JSON file"""
        default_config = {
            "email": {
                "smtp_server": "smtp.gmail.com",
                "smtp_port": 587,
                "sender_email": "",
                "sender_password": "",
                "receiver_emails": []
            },
            "trading": {
                "max_positions": 10,
                "position_size": 0.02,  # 2% per position
                "risk_per_trade": 0.01,  # 1% risk per trade
                "min_volume_ratio": 1.5,
                "min_confidence": 75
            },
            "patterns": {
                "min_pattern_length": 20,
                "max_pattern_length": 100,
                "volatility_threshold": 0.02
            }
        }

        try:
            with open(config_file, 'r') as f:
                config = json.load(f)
            return {**default_config, **config}
        except FileNotFoundError:
            with open(config_file, 'w') as f:
                json.dump(default_config, f, indent=2)
            return default_config

    def setup_database(self):
        """Setup SQLite database for storing historical data"""
        self.conn = sqlite3.connect('stock_patterns.db', check_same_thread=False)
        cursor = self.conn.cursor()

        # Create tables
        cursor.execute('''
            CREATE TABLE IF NOT EXISTS patterns (
                id INTEGER PRIMARY KEY AUTOINCREMENT,
                symbol TEXT,
                pattern_type TEXT,
                confidence REAL,
                detected_date TEXT,
                entry_price REAL,
                target_price REAL,
                stop_loss REAL,
                outcome TEXT,
                pnl REAL
            )
        ''')

        cursor.execute('''
            CREATE TABLE IF NOT EXISTS market_data (
                symbol TEXT,
                date TEXT,
                open REAL,
                high REAL,
                low REAL,
                close REAL,
                volume INTEGER,
                PRIMARY KEY (symbol, date)
            )
        ''')

        self.conn.commit()

    def load_nse_universe(self):
        """Load NSE stock universe"""
        try:
            # Top NSE stocks by market cap
            # nse_top_stocks = [
            #     'RELIANCE.NS', 'TCS.NS', 'HDFCBANK.NS', 'INFY.NS', 'HINDUNILVR.NS',
            #     'ICICIBANK.NS', 'SBIN.NS', 'BHARTIARTL.NS', 'LICI.NS', 'ITC.NS',
            #     'LT.NS', 'HCLTECH.NS', 'MARUTI.NS', 'SUNPHARMA.NS', 'TITAN.NS',
            #     'ONGC.NS', 'TATAMOTORS.NS', 'WIPRO.NS', 'ASIANPAINT.NS', 'ADANIENT.NS',
            #     'COALINDIA.NS', 'BAJFINANCE.NS', 'NESTLEIND.NS', 'POWERGRID.NS',
            #     'NTPC.NS', 'TECHM.NS', 'JSWSTEEL.NS', 'HINDALCO.NS', 'INDUSINDBK.NS',
            #     'HEROMOTOCO.NS', 'BRITANNIA.NS', 'BAJAJFINSV.NS', 'CIPLA.NS',
            #     'ULTRACEMCO.NS', 'DRREDDY.NS', 'EICHERMOT.NS', 'GRASIM.NS',
            #     'AXISBANK.NS', 'KOTAKBANK.NS', 'DIVISLAB.NS', 'TATACONSUM.NS',
            #     'BAJAJ-AUTO.NS', 'APOLLOHOSP.NS', 'ADANIPORTS.NS', 'SHREECEM.NS',
            #     'TATASTEEL.NS', 'BPCL.NS', 'IOC.NS', 'PIDILITIND.NS'
            # ]
            nse_top_stocks = [
              'AARTIIND.NS',      # Aarti Industries
              'ABB.NS',           # ABB India
              'ABBOTINDIA.NS',    # Abbott India
              'ABCAPITAL.NS',     # Aditya Birla Capital
              'ABFRL.NS',         # Aditya Birla Fashion & Retail
              'ACC.NS',           # ACC Limited
              'ADANIENT.NS',      # Adani Enterprises
              'ADANIPORTS.NS',    # Adani Ports & SEZ
              'ALKEM.NS',         # Alkem Laboratories
              'AMBUJACEM.NS',     # Ambuja Cements
              'APOLLOHOSP.NS',    # Apollo Hospitals
              'APOLLOTYRE.NS',    # Apollo Tyres
              'ASHOKLEY.NS',      # Ashok Leyland
              'ASIANPAINT.NS',    # Asian Paints
              'ASTRAL.NS',        # Astral Limited
              'ATUL.NS',          # Atul Limited
              'AUBANK.NS',        # AU Small Finance Bank
              'AUROPHARMA.NS',    # Aurobindo Pharma
              'AXISBANK.NS',      # Axis Bank
              'BAJAJ-AUTO.NS',    # Bajaj Auto
              'BAJAJFINSV.NS',    # Bajaj Finserv
              'BAJFINANCE.NS',    # Bajaj Finance
              'BALKRISIND.NS',    # Balkrishna Industries
              'BALRAMCHIN.NS',    # Balrampur Chini Mills
              'BANDHANBNK.NS',    # Bandhan Bank
              'BANKBARODA.NS',    # Bank of Baroda
              'BATAINDIA.NS',     # Bata India
              'BEL.NS',           # Bharat Electronics
              'BERGEPAINT.NS',    # Berger Paints
              'BHARATFORG.NS',    # Bharat Forge
              'BHARTIARTL.NS',    # Bharti Airtel
              'BHEL.NS',          # Bharat Heavy Electricals
              'BIOCON.NS',        # Biocon
              'BOSCHLTD.NS',      # Bosch Limited
              'BPCL.NS',          # Bharat Petroleum
              'BRITANNIA.NS',     # Britannia Industries
              'BSOFT.NS',         # Birlasoft
              'CANBK.NS',         # Canara Bank
              'CANFINHOME.NS',    # Can Fin Homes
              'CHAMBLFERT.NS',    # Chambal Fertilizers
              'CHOLAFIN.NS',      # Cholamandalam Investment
              'CIPLA.NS',         # Cipla
              'COALINDIA.NS',     # Coal India
              'COFORGE.NS',       # Coforge
              'COLPAL.NS',        # Colgate Palmolive
              'CONCOR.NS',        # Container Corporation
              'COROMANDEL.NS',    # Coromandel International
              'CROMPTON.NS',      # Crompton Greaves
              'CUB.NS',           # City Union Bank
              'CUMMINSIND.NS',    # Cummins India
              'DABUR.NS',         # Dabur India
              'DALBHARAT.NS',     # Dalmia Bharat
              'DEEPAKNTR.NS',     # Deepak Nitrite
              'DIVISLAB.NS',      # Divi's Laboratories
              'DLF.NS',           # DLF Limited
              'DMART.NS',         # Avenue Supermarts (DMart)
              'DRREDDY.NS',       # Dr. Reddy's Laboratories
              'EICHERMOT.NS',     # Eicher Motors
              'ESCORTS.NS',       # Escorts Limited
              'EXIDEIND.NS',      # Exide Industries
              'FEDERALBNK.NS',    # Federal Bank
              'GAIL.NS',          # GAIL India
              'GLENMARK.NS',      # Glenmark Pharmaceuticals
              'GMRINFRA.NS',      # GMR Infrastructure
              'GNFC.NS',          # Gujarat Narmada Valley Fertilizers
              'GODREJCP.NS',      # Godrej Consumer Products
              'GODREJPROP.NS',    # Godrej Properties
              'GRANULES.NS',      # Granules India
              'GRASIM.NS',        # Grasim Industries
              'GUJGASLTD.NS',     # Gujarat Gas
              'HAL.NS',           # Hindustan Aeronautics
              'HAVELLS.NS',       # Havells India
              'HCLTECH.NS',       # HCL Technologies
              'HDFC.NS',          # Housing Development Finance Corporation
              'HDFCAMC.NS',       # HDFC Asset Management
              'HDFCBANK.NS',      # HDFC Bank
              'HDFCLIFE.NS',      # HDFC Life Insurance
              'HEROMOTOCO.NS',    # Hero MotoCorp
              'HINDALCO.NS',      # Hindalco Industries
              'HINDCOPPER.NS',    # Hindustan Copper
              'HINDPETRO.NS',     # Hindustan Petroleum
              'HINDUNILVR.NS',    # Hindustan Unilever
              'HINDZINC.NS',      # Hindustan Zinc
              'HONAUT.NS',        # Honeywell Automation
              'HPCL.NS',          # Hindustan Petroleum Corporation
              'ICICIBANK.NS',     # ICICI Bank
              'ICICIGI.NS',       # ICICI Lombard General Insurance
              'ICICIPRULI.NS',    # ICICI Prudential Life Insurance
              'IDEA.NS',          # Vodafone Idea
              'IDFC.NS',          # IDFC Limited
              'IDFCFIRSTB.NS',    # IDFC First Bank
              'IEX.NS',           # Indian Energy Exchange
              'IGL.NS',           # Indraprastha Gas
              'INDHOTEL.NS',      # Indian Hotels
              'INDIGO.NS',        # InterGlobe Aviation
              'INDUSINDBK.NS',    # IndusInd Bank
              'INDUSTOWER.NS',    # Bharti Infratel
              'INFY.NS',          # Infosys
              'IOC.NS',           # Indian Oil Corporation
              'IPCALAB.NS',       # IPCA Laboratories
              'IRCTC.NS',         # Indian Railway Catering & Tourism Corporation
              'IRFC.NS',          # Indian Railway Finance Corporation
              'ITC.NS',           # ITC Limited
              'JINDALSTEL.NS',    # Jindal Steel & Power
              'JIOFIN.NS',        # Jio Financial Services
              'JKCEMENT.NS',      # JK Cement
              'JSWSTEEL.NS',      # JSW Steel
              'JUBLFOOD.NS',      # Jubilant Foodworks
              'KOTAKBANK.NS',     # Kotak Mahindra Bank
              'L&TFH.NS',         # L&T Finance Holdings
              'LALPATHLAB.NS',    # Dr. Lal PathLabs
              'LAURUSLABS.NS',    # Laurus Labs
              'LICHSGFIN.NS',     # LIC Housing Finance
              'LICI.NS',          # Life Insurance Corporation
              'LT.NS',            # Larsen & Toubro
              'LTI.NS',           # LTI Mindtree
              'LTTS.NS',          # L&T Technology Services
              'LUPIN.NS',         # Lupin Limited
              'M&M.NS',           # Mahindra & Mahindra
              'M&MFIN.NS',        # Mahindra & Mahindra Financial Services
              'MANAPPURAM.NS',    # Manappuram Finance
              'MARICO.NS',        # Marico Limited
              'MARUTI.NS',        # Maruti Suzuki
              'MCDOWELL-N.NS',    # United Spirits
              'MCX.NS',           # Multi Commodity Exchange
              'METROPOLIS.NS',    # Metropolis Healthcare
              'MFSL.NS',          # Max Financial Services
              'MGL.NS',           # Mahanagar Gas
              'MINDTREE.NS',      # Mindtree Limited
              'MOTHERSON.NS',     # Motherson Sumi Systems
              'MPHASIS.NS',       # Mphasis Limited
              'MRF.NS',           # MRF Limited
              'MUTHOOTFIN.NS',    # Muthoot Finance
              'NATIONALUM.NS',    # National Aluminium Company
              'NAUKRI.NS',        # Info Edge India
              'NAVINFLUOR.NS',    # Navin Fluorine International
              'NBCC.NS',          # NBCC India
              'NESTLEIND.NS',     # Nestle India
              'NMDC.NS',          # NMDC Limited
              'NTPC.NS',          # NTPC Limited
              'OBEROIRLTY.NS',    # Oberoi Realty
              'OFSS.NS',          # Oracle Financial Services Software
              'ONGC.NS',          # Oil & Natural Gas Corporation
              'PAGEIND.NS',       # Page Industries
              'PEL.NS',           # Piramal Enterprises
              'PERSISTENT.NS',    # Persistent Systems
              'PETRONET.NS',      # Petronet LNG
              'PFC.NS',           # Power Finance Corporation
              'PIDILITIND.NS',    # Pidilite Industries
              'PIIND.NS',         # Piramal Enterprises
              'PNB.NS',           # Punjab National Bank
              'POLYCAB.NS',       # Polycab India
              'POWERGRID.NS',     # Power Grid Corporation
              'PVRINOX.NS',       # PVR INOX
              'RAMCOCEM.NS',      # Ramco Cements
              'RBLBANK.NS',       # RBL Bank
              'RECLTD.NS',        # Rural Electrification Corporation
              'RELIANCE.NS',      # Reliance Industries
              'SAIL.NS',          # Steel Authority of India
              'SBICARD.NS',       # SBI Cards & Payment Services
              'SBILIFE.NS',       # SBI Life Insurance
              'SBIN.NS',          # State Bank of India
              'SHREECEM.NS',      # Shree Cement
              'SIEMENS.NS',       # Siemens Limited
              'SRF.NS',           # SRF Limited
              'SUNPHARMA.NS',     # Sun Pharmaceutical Industries
              'SUNTV.NS',         # Sun TV Network
              'SYNGENE.NS',       # Syngene International
              'TATACHEM.NS',      # Tata Chemicals
              'TATACOMM.NS',      # Tata Communications
              'TATACONSUM.NS',    # Tata Consumer Products
              'TATAMOTORS.NS',    # Tata Motors
              'TATAPOWER.NS',     # Tata Power
              'TATASTEEL.NS',     # Tata Steel
              'TCS.NS',           # Tata Consultancy Services
              'TECHM.NS',         # Tech Mahindra
              'TITAN.NS',         # Titan Company
              'TORNTPHARM.NS',    # Torrent Pharmaceuticals
              'TRENT.NS',         # Trent Limited
              'TVSMOTOR.NS',      # TVS Motor Company
              'UBL.NS',           # United Breweries
              'ULTRACEMCO.NS',    # UltraTech Cement
              'UPL.NS',           # UPL Limited
              'VEDL.NS',          # Vedanta Limited
              'VOLTAS.NS',        # Voltas Limited
              'WIPRO.NS',         # Wipro Limited
              'ZEEL.NS',          # Zee Entertainment Enterprises
              'ZOMATO.NS',        # Zomato Limited
              'ZYDUSLIFE.NS',     # Zydus Lifesciences
          ]
            return nse_top_stocks
        except Exception as e:
            logging.error(f"Error loading NSE universe: {e}")
            return []

    def load_bse_universe(self):
        """Load BSE stock universe"""
        return [stock.replace('.NS', '.BO') for stock in self.load_nse_universe()]

    def get_enhanced_stock_data(self, symbol, period='2y'):
        """Enhanced stock data fetching with multiple sources"""
        try:
            # Primary source: yfinance
            stock = yf.Ticker(symbol)
            data = stock.history(period=period)

            if data.empty:
                logging.warning(f"No data from yfinance for {symbol}")
                return None

            # Add enhanced features
            data = self.add_advanced_technical_indicators(data)
            data = self.add_market_microstructure_features(data)

            return data

        except Exception as e:
            logging.error(f"Error fetching enhanced data for {symbol}: {e}")
            return None

    def add_advanced_technical_indicators(self, data):
        """Add comprehensive technical indicators"""
        try:
            # Basic indicators
            data['RSI'] = ta.momentum.RSIIndicator(data['Close'], window=14).rsi()
            data['RSI_2'] = ta.momentum.RSIIndicator(data['Close'], window=2).rsi()
            data['MACD'] = ta.trend.MACD(data['Close']).macd()
            data['MACD_Signal'] = ta.trend.MACD(data['Close']).macd_signal()
            data['MACD_Histogram'] = data['MACD'] - data['MACD_Signal']

            # Multiple timeframe EMAs
            for period in [8, 13, 21, 34, 55, 89, 144]:
                data[f'EMA_{period}'] = ta.trend.EMAIndicator(data['Close'], window=period).ema_indicator()

            # Bollinger Bands with multiple periods
            for period in [20, 50]:
                bb = ta.volatility.BollingerBands(data['Close'], window=period)
                data[f'BB_Upper_{period}'] = bb.bollinger_hband()
                data[f'BB_Lower_{period}'] = bb.bollinger_lband()
                data[f'BB_Width_{period}'] = (data[f'BB_Upper_{period}'] - data[f'BB_Lower_{period}']) / data['Close']
                data[f'BB_Position_{period}'] = (data['Close'] - data[f'BB_Lower_{period}']) / (data[f'BB_Upper_{period}'] - data[f'BB_Lower_{period}'])

            # Volume indicators
            data['Volume_SMA_20'] = data['Volume'].rolling(window=20).mean()
            data['Volume_SMA_50'] = data['Volume'].rolling(window=50).mean()
            data['Volume_Ratio'] = data['Volume'] / data['Volume_SMA_20']
            data['OBV'] = ta.volume.OnBalanceVolumeIndicator(data['Close'], data['Volume']).on_balance_volume()
            data['VWAP'] = (data['Volume'] * (data['High'] + data['Low'] + data['Close']) / 3).cumsum() / data['Volume'].cumsum()

            # Momentum indicators
            data['ADX'] = ta.trend.ADXIndicator(data['High'], data['Low'], data['Close']).adx()
            data['CCI'] = ta.trend.CCIIndicator(data['High'], data['Low'], data['Close']).cci()
            data['Williams_R'] = ta.momentum.WilliamsRIndicator(data['High'], data['Low'], data['Close']).williams_r()
            data['Stoch_K'] = ta.momentum.StochasticOscillator(data['High'], data['Low'], data['Close']).stoch()
            data['Stoch_D'] = ta.momentum.StochasticOscillator(data['High'], data['Low'], data['Close']).stoch_signal()

            # Volatility indicators
            data['ATR'] = ta.volatility.AverageTrueRange(data['High'], data['Low'], data['Close']).average_true_range()
            data['Volatility'] = data['Close'].pct_change().rolling(window=20).std() * np.sqrt(252)

            # Support and Resistance
            data['Resistance'] = data['High'].rolling(window=20).max()
            data['Support'] = data['Low'].rolling(window=20).min()

            # Price patterns
            data['Higher_High'] = (data['High'] > data['High'].shift(1)) & (data['High'].shift(1) > data['High'].shift(2))
            data['Lower_Low'] = (data['Low'] < data['Low'].shift(1)) & (data['Low'].shift(1) < data['Low'].shift(2))

            return data

        except Exception as e:
            logging.error(f"Error adding technical indicators: {e}")
            return data

    def add_market_microstructure_features(self, data):
        """Add market microstructure and behavioral features"""
        try:
            # Price action features
            data['Body_Size'] = abs(data['Close'] - data['Open']) / data['Open']
            data['Upper_Shadow'] = (data['High'] - np.maximum(data['Open'], data['Close'])) / data['Open']
            data['Lower_Shadow'] = (np.minimum(data['Open'], data['Close']) - data['Low']) / data['Open']
            data['Range'] = (data['High'] - data['Low']) / data['Open']

            # Gap analysis
            data['Gap'] = (data['Open'] - data['Close'].shift(1)) / data['Close'].shift(1)
            data['Gap_Up'] = data['Gap'] > 0.02
            data['Gap_Down'] = data['Gap'] < -0.02

            # Momentum features
            for period in [1, 3, 5, 10]:
                data[f'Return_{period}d'] = data['Close'].pct_change(period)
                data[f'Vol_Adj_Return_{period}d'] = data[f'Return_{period}d'] / data['Volatility']

            # Regime detection
            data['Trend_Strength'] = np.where(
                data['Close'] > data['EMA_21'],
                (data['Close'] - data['EMA_21']) / data['EMA_21'],
                (data['Close'] - data['EMA_21']) / data['EMA_21']
            )

            # Volume profile approximation
            data['Volume_Profile'] = data['Volume'] / data['Volume'].rolling(window=50).max()

            return data

        except Exception as e:
            logging.error(f"Error adding microstructure features: {e}")
            return data

    def detect_advanced_patterns(self, data, symbol):
        """Detect advanced chart patterns using ML and statistical methods"""
        patterns = []

        try:
            # Traditional patterns with ML enhancement
            patterns.extend(self.detect_harmonic_patterns(data))
            patterns.extend(self.detect_elliott_wave_patterns(data))
            patterns.extend(self.detect_volume_patterns(data))
            patterns.extend(self.detect_breakout_patterns(data))
            patterns.extend(self.detect_reversal_patterns(data))
            patterns.extend(self.detect_continuation_patterns(data))
            patterns.extend(self.detect_anomaly_patterns(data))

            # ML-based pattern detection
            if self.pattern_classifier:
                ml_patterns = self.detect_ml_patterns(data)
                patterns.extend(ml_patterns)

        except Exception as e:
            logging.error(f"Error detecting patterns for {symbol}: {e}")

        return patterns

    def detect_harmonic_patterns(self, data):
        """Detect harmonic patterns (Gartley, Butterfly, Bat, Crab)"""
        patterns = []

        try:
            if len(data) < 100:
                return patterns

            # Simplified harmonic pattern detection
            closes = data['Close'].values
            highs = data['High'].values
            lows = data['Low'].values

            # Find significant pivots
            high_pivots = find_peaks(highs, distance=10, prominence=np.std(highs)*0.5)[0]
            low_pivots = find_peaks(-lows, distance=10, prominence=np.std(lows)*0.5)[0]

            if len(high_pivots) >= 3 and len(low_pivots) >= 2:
                # Check for potential Gartley pattern
                pattern = self.validate_gartley_pattern(data, high_pivots, low_pivots)
                if pattern:
                    patterns.append(pattern)

        except Exception as e:
            logging.error(f"Error detecting harmonic patterns: {e}")

        return patterns

    def validate_gartley_pattern(self, data, high_pivots, low_pivots):
        """Validate Gartley harmonic pattern"""
        try:
            # Get recent pivots
            if len(high_pivots) < 2 or len(low_pivots) < 2:
                return None

            # Simplified Gartley validation
            current_price = data['Close'].iloc[-1]

            # Check Fibonacci retracements (simplified)
            recent_high = data['High'].iloc[high_pivots[-1]]
            recent_low = data['Low'].iloc[low_pivots[-1]]

            retracement = (current_price - recent_low) / (recent_high - recent_low)

            if 0.618 <= retracement <= 0.786:  # Gartley B point retracement
                return {
                    'pattern': 'Gartley Pattern',
                    'type': 'bullish' if current_price < recent_high else 'bearish',
                    'confidence': 80,
                    'entry_level': current_price,
                    'target': recent_high if current_price < recent_high else recent_low,
                    'retracement_level': retracement
                }

        except Exception as e:
            logging.error(f"Error validating Gartley pattern: {e}")

        return None

    def detect_elliott_wave_patterns(self, data):
        """Detect Elliott Wave patterns"""
        patterns = []

        try:
            if len(data) < 50:
                return patterns

            # Simplified Elliott Wave detection
            closes = data['Close'].values

            # Find wave structure using pivot points
            pivots = self.find_elliott_pivots(closes)

            if len(pivots) >= 5:
                wave_pattern = self.analyze_wave_structure(pivots, closes)
                if wave_pattern:
                    patterns.append(wave_pattern)

        except Exception as e:
            logging.error(f"Error detecting Elliott Wave patterns: {e}")

        return patterns

    def find_elliott_pivots(self, prices):
        """Find pivots for Elliott Wave analysis"""
        high_pivots = find_peaks(prices, distance=5)[0]
        low_pivots = find_peaks(-prices, distance=5)[0]

        # Combine and sort pivots
        all_pivots = list(high_pivots) + list(low_pivots)
        all_pivots.sort()

        return all_pivots

    def analyze_wave_structure(self, pivots, prices):
        """Analyze Elliott Wave structure"""
        try:
            if len(pivots) < 5:
                return None

            # Get recent 5 waves
            recent_pivots = pivots[-5:]
            wave_prices = [prices[i] for i in recent_pivots]

            # Simple wave validation (5-wave impulse pattern)
            if len(wave_prices) == 5:
                # Check for alternation and extension
                wave1 = abs(wave_prices[1] - wave_prices[0])
                wave3 = abs(wave_prices[3] - wave_prices[2])
                wave5 = abs(wave_prices[4] - wave_prices[3])

                # Wave 3 should be longest (simplified rule)
                if wave3 > wave1 and wave3 > wave5:
                    return {
                        'pattern': 'Elliott Wave Impulse',
                        'type': 'trending',
                        'confidence': 70,
                        'wave_count': 5,
                        'current_wave': 5
                    }

        except Exception as e:
            logging.error(f"Error analyzing wave structure: {e}")

        return None

    def detect_volume_patterns(self, data):
        """Detect volume-based patterns"""
        patterns = []

        try:
            # Volume breakout pattern
            avg_volume = data['Volume'].rolling(window=20).mean()
            volume_spike = data['Volume'] > (avg_volume * 2)
            price_breakout = data['Close'] > data['Close'].rolling(window=20).max().shift(1)

            if volume_spike.iloc[-1] and price_breakout.iloc[-1]:
                patterns.append({
                    'pattern': 'Volume Breakout',
                    'type': 'bullish',
                    'confidence': 85,
                    'volume_ratio': data['Volume'].iloc[-1] / avg_volume.iloc[-1],
                    'breakout_level': data['Close'].rolling(window=20).max().shift(1).iloc[-1]
                })

            # Volume divergence pattern
            price_trend = data['Close'].rolling(window=10).apply(lambda x: stats.linregress(range(len(x)), x)[0])
            volume_trend = data['Volume'].rolling(window=10).apply(lambda x: stats.linregress(range(len(x)), x)[0])

            if (price_trend.iloc[-1] > 0 and volume_trend.iloc[-1] < 0) or \
               (price_trend.iloc[-1] < 0 and volume_trend.iloc[-1] > 0):
                patterns.append({
                    'pattern': 'Volume Divergence',
                    'type': 'warning',
                    'confidence': 75,
                    'price_trend': price_trend.iloc[-1],
                    'volume_trend': volume_trend.iloc[-1]
                })

        except Exception as e:
            logging.error(f"Error detecting volume patterns: {e}")

        return patterns

    def detect_breakout_patterns(self, data):
        """Detect breakout patterns with confirmation"""
        patterns = []

        try:
            # Range breakout
            range_high = data['High'].rolling(window=20).max()
            range_low = data['Low'].rolling(window=20).min()
            range_size = (range_high - range_low) / data['Close']

            # Look for consolidation followed by breakout
            consolidation = range_size < 0.15  # Less than 15% range
            breakout_up = data['Close'] > range_high.shift(1)
            breakout_down = data['Close'] < range_low.shift(1)

            if consolidation.rolling(window=10).sum().iloc[-1] >= 7:  # 7 out of 10 days consolidating
                if breakout_up.iloc[-1]:
                    patterns.append({
                        'pattern': 'Range Breakout (Bullish)',
                        'type': 'bullish',
                        'confidence': 80,
                        'breakout_level': range_high.iloc[-2],
                        'target': range_high.iloc[-2] + (range_high.iloc[-2] - range_low.iloc[-2]) * 0.618
                    })
                elif breakout_down.iloc[-1]:
                    patterns.append({
                        'pattern': 'Range Breakout (Bearish)',
                        'type': 'bearish',
                        'confidence': 80,
                        'breakout_level': range_low.iloc[-2],
                        'target': range_low.iloc[-2] - (range_high.iloc[-2] - range_low.iloc[-2]) * 0.618
                    })

        except Exception as e:
            logging.error(f"Error detecting breakout patterns: {e}")

        return patterns

    def detect_reversal_patterns(self, data):
        """Detect reversal patterns"""
        patterns = []

        try:
            # Hammer/Doji patterns
            for i in range(-5, 0):  # Check last 5 candles
                if abs(i) > len(data):
                    continue

                candle = data.iloc[i]
                body_size = abs(candle['Close'] - candle['Open']) / candle['Open']
                upper_shadow = (candle['High'] - max(candle['Open'], candle['Close'])) / candle['Open']
                lower_shadow = (min(candle['Open'], candle['Close']) - candle['Low']) / candle['Open']

                # Hammer pattern
                if (body_size < 0.01 and lower_shadow > 0.02 and upper_shadow < 0.005):
                    patterns.append({
                        'pattern': 'Hammer',
                        'type': 'bullish',
                        'confidence': 70,
                        'candle_index': i,
                        'body_size': body_size,
                        'lower_shadow': lower_shadow
                    })

                # Shooting star pattern
                if (body_size < 0.01 and upper_shadow > 0.02 and lower_shadow < 0.005):
                    patterns.append({
                        'pattern': 'Shooting Star',
                        'type': 'bearish',
                        'confidence': 70,
                        'candle_index': i,
                        'body_size': body_size,
                        'upper_shadow': upper_shadow
                    })

        except Exception as e:
            logging.error(f"Error detecting reversal patterns: {e}")

        return patterns

    def detect_continuation_patterns(self, data):
        """Detect continuation patterns"""
        patterns = []

        try:
            # Flag pattern
            if len(data) >= 30:
                # Look for strong move followed by consolidation
                price_change_20 = (data['Close'].iloc[-20] - data['Close'].iloc[-30]) / data['Close'].iloc[-30]
                recent_volatility = data['Close'].iloc[-10:].pct_change().std()

                if abs(price_change_20) > 0.1 and recent_volatility < 0.02:  # 10% move, then low volatility
                    pattern_type = 'bullish' if price_change_20 > 0 else 'bearish'
                    patterns.append({
                        'pattern': f'Flag Pattern ({pattern_type.title()})',
                        'type': pattern_type,
                        'confidence': 75,
                        'pole_size': abs(price_change_20),
                        'consolidation_volatility': recent_volatility
                    })

        except Exception as e:
            logging.error(f"Error detecting continuation patterns: {e}")

        return patterns

    def detect_anomaly_patterns(self, data):
        """Detect anomalous patterns using statistical methods"""
        patterns = []

        try:
            # Prepare features for anomaly detection
            features = []
            for col in ['Close', 'Volume', 'RSI', 'MACD']:
                if col in data.columns:
                    features.append(data[col].fillna(method='ffill').values[-50:])  # Last 50 days

            if features:
                features_array = np.column_stack(features)

                # Detect anomalies
                anomalies = self.anomaly_detector.fit_predict(features_array)

                # If recent data points are anomalies
                if anomalies[-1] == -1:  # -1 indicates anomaly
                    patterns.append({
                        'pattern': 'Statistical Anomaly',
                        'type': 'warning',
                        'confidence': 60,
                        'anomaly_score': self.anomaly_detector.score_samples(features_array)[-1]
                    })

        except Exception as e:
            logging.error(f"Error detecting anomaly patterns: {e}")

        return patterns

    def detect_ml_patterns(self, data):
        """Use ML model to detect patterns"""
        patterns = []

        try:
            if self.pattern_classifier and len(data) >= 50:
                # Prepare features
                features = self.prepare_ml_features(data)

                if features is not None:
                    # Predict pattern
                    prediction = self.pattern_classifier.predict([features])
                    confidence = self.pattern_classifier.predict_proba([features]).max()

                    if confidence > 0.7:  # High confidence threshold
                        pattern_names = ['No Pattern', 'Bullish', 'Bearish', 'Neutral']
                        patterns.append({
                            'pattern': f'ML Pattern: {pattern_names[prediction[0]]}',
                            'type': pattern_names[prediction[0]].lower(),
                            'confidence': confidence * 100,
                            'ml_prediction': True
                        })

        except Exception as e:
            logging.error(f"Error with ML pattern detection: {e}")

        return patterns

    def prepare_ml_features(self, data):
        """Prepare features for ML model"""
        try:
            features = []

            # Recent price action features
            features.extend([
                data['RSI'].iloc[-1] if 'RSI' in data.columns else 50,
                data['MACD'].iloc[-1] if 'MACD' in data.columns else 0,
                data['Volume_Ratio'].iloc[-1] if 'Volume_Ratio' in data.columns else 1,
                data['ATR'].iloc[-1] if 'ATR' in data.columns else 0.02,
                data['BB_Position_20'].iloc[-1] if 'BB_Position_20' in data.columns else 0.5
            ])

            # Price momentum features
            if len(data) >= 10:
                features.extend([
                    data['Close'].pct_change(5).iloc[-1],
                    data['Close'].pct_change(10).iloc[-1],
                    data['Volume'].rolling(5).mean().iloc[-1] / data['Volume'].rolling(20).mean().iloc[-1]
                ])

            return np.array(features) if len(features) > 0 else None

        except Exception as e:
            logging.error(f"Error preparing ML features: {e}")
            return None

    def initialize_models(self):
        """Initialize ML models"""
        try:
            # Initialize pattern classifier with dummy data
            X_dummy = np.random.rand(100, 8)  # 8 features
            y_dummy = np.random.randint(0, 4, 100)  # 4 classes

            self.pattern_classifier = RandomForestClassifier(
                n_estimators=100,
                max_depth=10,
                random_state=42
            )
            self.pattern_classifier.fit(X_dummy, y_dummy)

            logging.info("ML models initialized successfully")

        except Exception as e:
            logging.error(f"Error initializing ML models: {e}")

    def calculate_position_size(self, signal, account_balance=100000):
        """Calculate optimal position size using Kelly Criterion"""
        try:
            # Kelly Criterion: f = (bp - q) / b
            # where f = fraction of capital to wager
            # b = odds received on the wager (target/risk)
            # p = probability of winning
            # q = probability of losing (1 - p)

            win_prob = signal.confidence / 100
            loss_prob = 1 - win_prob

            if signal.target_price and signal.stop_loss:
                # Calculate risk-reward ratio
                risk = abs(signal.entry_price - signal.stop_loss) / signal.entry_price
                reward = abs(signal.target_price - signal.entry_price) / signal.entry_price

                if risk > 0:
                    odds = reward / risk
                    kelly_fraction = (win_prob * odds - loss_prob) / odds

                    # Apply Kelly with safety factor (25% of Kelly)
                    kelly_fraction = max(0, min(kelly_fraction * 0.25, 0.05))  # Max 5% per trade

                    position_size = account_balance * kelly_fraction

                    return {
                        'position_size': position_size,
                        'kelly_fraction': kelly_fraction,
                        'risk_reward_ratio': odds,
                        'risk_amount': position_size * risk
                    }

            # Default position sizing (2% of account)
            return {
                'position_size': account_balance * 0.02,
                'kelly_fraction': 0.02,
                'risk_reward_ratio': 1.0,
                'risk_amount': account_balance * 0.02 * 0.02
            }

        except Exception as e:
            logging.error(f"Error calculating position size: {e}")
            return {'position_size': account_balance * 0.01, 'kelly_fraction': 0.01}

    def get_market_sentiment(self, symbol):
        """Get market sentiment from multiple sources"""
        try:
            sentiment_score = 0.0

            # VIX-based market fear gauge (simplified)
            try:
                vix = yf.Ticker("^VIX")
                vix_data = vix.history(period="5d")
                if not vix_data.empty:
                    current_vix = vix_data['Close'].iloc[-1]
                    # Normalize VIX to sentiment score (-1 to 1)
                    sentiment_score += (25 - current_vix) / 25  # Inverted VIX
            except:
                pass

            # Dollar Index impact
            try:
                dxy = yf.Ticker("DX-Y.NYB")
                dxy_data = dxy.history(period="5d")
                if not dxy_data.empty:
                    dxy_change = dxy_data['Close'].pct_change(5).iloc[-1]
                    sentiment_score -= dxy_change * 2  # Strong dollar = negative for Indian stocks
            except:
                pass

            # Crude oil impact (India is oil importer)
            try:
                oil = yf.Ticker("CL=F")
                oil_data = oil.history(period="5d")
                if not oil_data.empty:
                    oil_change = oil_data['Close'].pct_change(5).iloc[-1]
                    sentiment_score -= oil_change * 0.5  # Higher oil = negative for India
            except:
                pass

            # Normalize sentiment score
            sentiment_score = max(-1, min(1, sentiment_score))

            return {
                'sentiment_score': sentiment_score,
                'sentiment_label': self.get_sentiment_label(sentiment_score)
            }

        except Exception as e:
            logging.error(f"Error getting market sentiment: {e}")
            return {'sentiment_score': 0.0, 'sentiment_label': 'Neutral'}

    def get_sentiment_label(self, score):
        """Convert sentiment score to label"""
        if score > 0.3:
            return 'Very Bullish'
        elif score > 0.1:
            return 'Bullish'
        elif score > -0.1:
            return 'Neutral'
        elif score > -0.3:
            return 'Bearish'
        else:
            return 'Very Bearish'

    def get_market_regime(self, symbol):
        """Determine current market regime"""
        try:
            # Get NIFTY data for market regime
            nifty = yf.Ticker("^NSEI")
            nifty_data = nifty.history(period="6mo")

            if nifty_data.empty:
                return 'Unknown'

            # Calculate regime indicators
            returns = nifty_data['Close'].pct_change().dropna()
            volatility = returns.rolling(window=20).std() * np.sqrt(252)
            trend = nifty_data['Close'].rolling(window=50).apply(lambda x: stats.linregress(range(len(x)), x)[0])

            current_vol = volatility.iloc[-1]
            current_trend = trend.iloc[-1]

            # Determine regime
            if current_vol > 0.25:  # High volatility
                if abs(current_trend) > 0.1:
                    return 'High Volatility Trending'
                else:
                    return 'High Volatility Ranging'
            else:  # Low volatility
                if current_trend > 0.05:
                    return 'Low Volatility Bull Market'
                elif current_trend < -0.05:
                    return 'Low Volatility Bear Market'
                else:
                    return 'Low Volatility Ranging'

        except Exception as e:
            logging.error(f"Error determining market regime: {e}")
            return 'Unknown'

    def generate_advanced_signal(self, symbol, patterns, data):
        """Generate advanced trading signals with comprehensive analysis"""
        try:
            if not patterns:
                return None

            # Get market context
            sentiment = self.get_market_sentiment(symbol)
            market_regime = self.get_market_regime(symbol)

            # Select best pattern
            best_pattern = max(patterns, key=lambda x: x.get('confidence', 0))

            current_price = data['Close'].iloc[-1]
            atr = data['ATR'].iloc[-1] if 'ATR' in data.columns else current_price * 0.02

            # Generate signal based on pattern type
            signal_type = 'WATCH'
            entry_price = current_price
            target_price = None
            stop_loss = None

            if best_pattern['type'] == 'bullish':
                # Bullish signal logic
                if sentiment['sentiment_score'] > -0.2:  # Not too bearish
                    signal_type = 'BUY'
                    entry_price = current_price

                    # Dynamic target based on ATR and pattern
                    if 'target' in best_pattern:
                        target_price = best_pattern['target']
                    else:
                        target_price = current_price * (1 + 2 * atr / current_price)  # 2 ATR target

                    # Dynamic stop loss
                    stop_loss = current_price * (1 - 1.5 * atr / current_price)  # 1.5 ATR stop

            elif best_pattern['type'] == 'bearish':
                # Bearish signal logic
                if sentiment['sentiment_score'] < 0.2:  # Not too bullish
                    signal_type = 'SELL'
                    entry_price = current_price

                    # Dynamic target based on ATR and pattern
                    if 'target' in best_pattern:
                        target_price = best_pattern['target']
                    else:
                        target_price = current_price * (1 - 2 * atr / current_price)  # 2 ATR target

                    # Dynamic stop loss
                    stop_loss = current_price * (1 + 1.5 * atr / current_price)  # 1.5 ATR stop

            # Calculate risk-reward ratio
            risk_reward_ratio = 0
            if target_price and stop_loss:
                risk = abs(entry_price - stop_loss)
                reward = abs(target_price - entry_price)
                risk_reward_ratio = reward / risk if risk > 0 else 0

            # Volume confirmation
            volume_confirmation = False
            if 'Volume_Ratio' in data.columns:
                volume_confirmation = data['Volume_Ratio'].iloc[-1] > 1.5

            # Create advanced signal
            signal = TradingSignal(
                symbol=symbol.replace('.NS', ''),
                pattern=best_pattern['pattern'],
                signal_type=signal_type,
                confidence=min(best_pattern['confidence'] *
                             (1 + sentiment['sentiment_score'] * 0.1), 100),  # Sentiment adjustment
                entry_price=entry_price,
                target_price=target_price,
                stop_loss=stop_loss,
                risk_reward_ratio=risk_reward_ratio,
                volume_confirmation=volume_confirmation,
                sentiment_score=sentiment['sentiment_score'],
                market_regime=market_regime,
                timestamp=datetime.now()
            )

            return signal

        except Exception as e:
            logging.error(f"Error generating advanced signal for {symbol}: {e}")
            return None

    def scan_stocks_parallel(self, max_workers=5):
        """Scan stocks in parallel for better performance"""
        results = []

        with ThreadPoolExecutor(max_workers=max_workers) as executor:
            # Submit all tasks
            future_to_symbol = {
                executor.submit(self.analyze_single_stock, symbol): symbol
                for symbol in self.nse_stocks[:50]  # Top 50 stocks
            }

            # Collect results
            for future in as_completed(future_to_symbol):
                symbol = future_to_symbol[future]
                try:
                    result = future.result()
                    if result:
                        results.append(result)
                        logging.info(f"‚úÖ Analyzed {symbol}")
                except Exception as e:
                    logging.error(f"‚ùå Error analyzing {symbol}: {e}")

        return results

    def analyze_single_stock(self, symbol):
        """Analyze a single stock comprehensively"""
        try:
            # Get enhanced stock data
            data = self.get_enhanced_stock_data(symbol)
            if data is None or len(data) < 50:
                return None

            # Detect patterns
            patterns = self.detect_advanced_patterns(data, symbol)
            if not patterns:
                return None

            # Generate signal
            signal = self.generate_advanced_signal(symbol, patterns, data)
            if not signal:
                return None

            # Calculate position sizing
            position_info = self.calculate_position_size(signal)

            # Store in database
            self.store_pattern_in_db(signal, patterns)

            return {
                'signal': signal,
                'patterns': patterns,
                'position_info': position_info,
                'technical_data': {
                    'rsi': data['RSI'].iloc[-1] if 'RSI' in data.columns else None,
                    'macd': data['MACD'].iloc[-1] if 'MACD' in data.columns else None,
                    'volume_ratio': data['Volume_Ratio'].iloc[-1] if 'Volume_Ratio' in data.columns else None,
                    'atr': data['ATR'].iloc[-1] if 'ATR' in data.columns else None
                }
            }

        except Exception as e:
            logging.error(f"Error in single stock analysis for {symbol}: {e}")
            return None

    def store_pattern_in_db(self, signal, patterns):
        """Store detected patterns in database"""
        try:
            cursor = self.conn.cursor()

            for pattern in patterns:
                cursor.execute('''
                    INSERT INTO patterns (symbol, pattern_type, confidence, detected_date,
                                        entry_price, target_price, stop_loss, outcome, pnl)
                    VALUES (?, ?, ?, ?, ?, ?, ?, ?, ?)
                ''', (
                    signal.symbol,
                    pattern['pattern'],
                    pattern.get('confidence', 0),
                    signal.timestamp.isoformat(),
                    signal.entry_price,
                    signal.target_price,
                    signal.stop_loss,
                    'PENDING',
                    0.0
                ))

            self.conn.commit()

        except Exception as e:
            logging.error(f"Error storing pattern in database: {e}")

    def create_advanced_visualization(self, symbol, data, patterns, signal):
        """Create advanced chart visualization"""
        try:
            fig = make_subplots(
                rows=4, cols=1,
                shared_xaxes=True,
                vertical_spacing=0.03,
                subplot_titles=('Price & Patterns', 'Volume', 'RSI', 'MACD'),
                row_heights=[0.5, 0.2, 0.15, 0.15]
            )

            # Candlestick chart
            fig.add_trace(
                go.Candlestick(
                    x=data.index,
                    open=data['Open'],
                    high=data['High'],
                    low=data['Low'],
                    close=data['Close'],
                    name='Price'
                ),
                row=1, col=1
            )

            # Add EMAs
            if 'EMA_21' in data.columns:
                fig.add_trace(
                    go.Scatter(x=data.index, y=data['EMA_21'], name='EMA 21',
                             line=dict(color='orange', width=1)),
                    row=1, col=1
                )

            if 'EMA_55' in data.columns:
                fig.add_trace(
                    go.Scatter(x=data.index, y=data['EMA_55'], name='EMA 55',
                             line=dict(color='blue', width=1)),
                    row=1, col=1
                )

            # Add signal markers
            if signal.signal_type in ['BUY', 'SELL']:
                color = 'green' if signal.signal_type == 'BUY' else 'red'
                fig.add_trace(
                    go.Scatter(
                        x=[data.index[-1]],
                        y=[signal.entry_price],
                        mode='markers',
                        marker=dict(size=15, color=color, symbol='triangle-up' if signal.signal_type == 'BUY' else 'triangle-down'),
                        name=f'{signal.signal_type} Signal'
                    ),
                    row=1, col=1
                )

            # Volume
            fig.add_trace(
                go.Bar(x=data.index, y=data['Volume'], name='Volume',
                      marker_color='lightblue'),
                row=2, col=1
            )

            # RSI
            if 'RSI' in data.columns:
                fig.add_trace(
                    go.Scatter(x=data.index, y=data['RSI'], name='RSI',
                             line=dict(color='purple')),
                    row=3, col=1
                )
                # RSI levels
                fig.add_hline(y=70, line_dash="dash", line_color="red", row=3, col=1)
                fig.add_hline(y=30, line_dash="dash", line_color="green", row=3, col=1)

            # MACD
            if 'MACD' in data.columns and 'MACD_Signal' in data.columns:
                fig.add_trace(
                    go.Scatter(x=data.index, y=data['MACD'], name='MACD',
                             line=dict(color='blue')),
                    row=4, col=1
                )
                fig.add_trace(
                    go.Scatter(x=data.index, y=data['MACD_Signal'], name='MACD Signal',
                             line=dict(color='red')),
                    row=4, col=1
                )

            fig.update_layout(
                title=f'{symbol} - {signal.pattern} ({signal.signal_type})',
                xaxis_rangeslider_visible=False,
                height=800
            )

            # Save chart
            chart_filename = f'charts/{symbol}_{datetime.now().strftime("%Y%m%d_%H%M%S")}.html'
            fig.write_html(chart_filename)

            return chart_filename

        except Exception as e:
            logging.error(f"Error creating visualization: {e}")
            return None

    def send_advanced_alert(self, analysis_result):
        """Send comprehensive alert with all information"""
        try:
            signal = analysis_result['signal']
            patterns = analysis_result['patterns']
            position_info = analysis_result['position_info']
            technical_data = analysis_result['technical_data']

            # Create comprehensive message
            message = f"""
üö® ADVANCED TRADING ALERT üö®
{'=' * 50}

üìä STOCK: {signal.symbol}
üéØ PATTERN: {signal.pattern}
üìà SIGNAL: {signal.signal_type}
üí™ CONFIDENCE: {signal.confidence:.1f}%
üß† MARKET REGIME: {signal.market_regime}
üòä SENTIMENT: {signal.sentiment_score:.2f} ({self.get_sentiment_label(signal.sentiment_score)})

üí∞ PRICE INFORMATION:
‚îú‚îÄ‚îÄ Current Price: ‚Çπ{signal.entry_price:.2f}
‚îú‚îÄ‚îÄ Target Price: ‚Çπ{signal.target_price:.2f} ({((signal.target_price/signal.entry_price-1)*100):.1f}%)
‚îú‚îÄ‚îÄ Stop Loss: ‚Çπ{signal.stop_loss:.2f} ({((signal.stop_loss/signal.entry_price-1)*100):.1f}%)
‚îî‚îÄ‚îÄ Risk:Reward = 1:{signal.risk_reward_ratio:.2f}

üìä TECHNICAL INDICATORS:
‚îú‚îÄ‚îÄ RSI: {technical_data.get('rsi', 'N/A'):.1f}
‚îú‚îÄ‚îÄ MACD: {technical_data.get('macd', 'N/A'):.4f}
‚îú‚îÄ‚îÄ Volume Ratio: {technical_data.get('volume_ratio', 'N/A'):.1f}x
‚îî‚îÄ‚îÄ ATR: {technical_data.get('atr', 'N/A'):.2f}

üíº POSITION SIZING:
‚îú‚îÄ‚îÄ Recommended Size: ‚Çπ{position_info['position_size']:,.0f}
‚îú‚îÄ‚îÄ Kelly Fraction: {position_info['kelly_fraction']:.2f}%
‚îú‚îÄ‚îÄ Risk Amount: ‚Çπ{position_info['risk_amount']:,.0f}
‚îî‚îÄ‚îÄ R:R Ratio: {position_info['risk_reward_ratio']:.2f}

üìã DETECTED PATTERNS:
{self.format_patterns_list(patterns)}

‚ö†Ô∏è CONFIRMATIONS:
‚îú‚îÄ‚îÄ Volume: {'‚úÖ' if signal.volume_confirmation else '‚ùå'}
‚îú‚îÄ‚îÄ Sentiment: {'‚úÖ' if signal.sentiment_score > 0 else '‚ùå'}
‚îî‚îÄ‚îÄ Market Regime: {signal.market_regime}

‚è∞ TIMESTAMP: {signal.timestamp.strftime('%Y-%m-%d %H:%M:%S')}

{'=' * 50}
‚ö†Ô∏è This is for educational purposes only. Trade responsibly!
            """

            # Send email alert
            self.send_email_alert_advanced(message, signal.symbol)

            # Log the alert
            logging.info(f"Advanced alert sent for {signal.symbol}")

            return message

        except Exception as e:
            logging.error(f"Error sending advanced alert: {e}")
            return None

    def format_patterns_list(self, patterns):
        """Format patterns list for display"""
        if not patterns:
            return "No patterns detected"

        formatted = ""
        for i, pattern in enumerate(patterns, 1):
            formatted += f"‚îú‚îÄ‚îÄ {i}. {pattern['pattern']} ({pattern.get('confidence', 0):.0f}%)\n"

        return formatted.rstrip('\n')

    def send_email_alert_advanced(self, message, symbol):
        """Send advanced email alert"""
        try:
            if not self.config['email']['sender_email']:
                return

            msg = MimeMultipart()
            msg['From'] = self.config['email']['sender_email']
            msg['To'] = ', '.join(self.config['email']['receiver_emails'])
            msg['Subject'] = f"üö® Advanced Stock Alert: {symbol}"

            msg.attach(MimeText(message, 'plain'))

            server = smtplib.SMTP(self.config['email']['smtp_server'], self.config['email']['smtp_port'])
            server.starttls()
            server.login(self.config['email']['sender_email'], self.config['email']['sender_password'])

            for receiver in self.config['email']['receiver_emails']:
                server.sendmail(self.config['email']['sender_email'], receiver, msg.as_string())

            server.quit()
            logging.info(f"Email alert sent successfully for {symbol}")

        except Exception as e:
            logging.error(f"Failed to send email alert: {e}")

    def run_advanced_scanner(self):
        """Run the advanced scanner with all features"""
        try:
            logging.info("üöÄ Starting Advanced Indian Stock Pattern Scanner...")

            # Scan stocks in parallel
            results = self.scan_stocks_parallel()

            if not results:
                logging.info("No significant patterns detected in this scan.")
                return

            # Sort by confidence
            results.sort(key=lambda x: x['signal'].confidence, reverse=True)

            logging.info(f"üéØ Found {len(results)} high-quality signals!")

            # Process each result
            for result in results:
                try:
                    # Send alert
                    alert_message = self.send_advanced_alert(result)

                    # Create visualization
                    symbol = result['signal'].symbol + '.NS'
                    data = self.get_enhanced_stock_data(symbol)
                    if data is not None:
                        chart_file = self.create_advanced_visualization(
                            symbol, data, result['patterns'], result['signal']
                        )
                        if chart_file:
                            logging.info(f"Chart saved: {chart_file}")

                    # Print summary
                    print(f"\n{'='*60}")
                    print(alert_message)
                    print(f"{'='*60}")

                except Exception as e:
                    logging.error(f"Error processing result: {e}")

            # Save results to file
            self.save_scan_results(results)

        except Exception as e:
            logging.error(f"Error in advanced scanner: {e}")

    def save_scan_results(self, results):
        """Save scan results to JSON file"""
        try:
            scan_data = {
                'timestamp': datetime.now().isoformat(),
                'total_signals': len(results),
                'signals': []
            }

            for result in results:
                signal = result['signal']
                scan_data['signals'].append({
                    'symbol': signal.symbol,
                    'pattern': signal.pattern,
                    'signal_type': signal.signal_type,
                    'confidence': signal.confidence,
                    'entry_price': signal.entry_price,
                    'target_price': signal.target_price,
                    'stop_loss': signal.stop_loss,
                    'risk_reward_ratio': signal.risk_reward_ratio,
                    'sentiment_score': signal.sentiment_score,
                    'market_regime': signal.market_regime,
                    'position_size': result['position_info']['position_size']
                })

            with open(f'scan_results_{datetime.now().strftime("%Y%m%d_%H%M%S")}.json', 'w') as f:
                json.dump(scan_data, f, indent=2)

            logging.info("Scan results saved successfully")

        except Exception as e:
            logging.error(f"Error saving scan results: {e}")

    def run_continuous_advanced_scanner(self, interval_minutes=30):
        """Run continuous advanced scanning"""
        def scan_job():
            try:
                logging.info(f"‚è∞ Starting scheduled scan at {datetime.now()}")
                self.run_advanced_scanner()
            except Exception as e:
                logging.error(f"Error in scheduled scan: {e}")

        # Schedule scans
        schedule.every(interval_minutes).minutes.do(scan_job)

        logging.info(f"üîÑ Continuous scanner started (every {interval_minutes} minutes)")

        # Run initial scan
        scan_job()

        # Keep running
        while True:
            try:
                schedule.run_pending()
                time.sleep(60)  # Check every minute
            except KeyboardInterrupt:
                logging.info("üõë Stopping continuous scanner...")
                break
            except Exception as e:
                logging.error(f"Error in continuous scanner: {e}")
                time.sleep(300)  # Wait 5 minutes before retrying

# Main execution
if __name__ == "__main__":
    # Create necessary directories
    import os
    os.makedirs('charts', exist_ok=True)

    # Initialize the advanced detector
    detector = AdvancedStockPatternDetector()

    print("üáÆüá≥ ADVANCED INDIAN STOCK PATTERN DETECTION SYSTEM")
    print("=" * 60)
    print("üöÄ Features:")
    print("‚îú‚îÄ‚îÄ Advanced Pattern Recognition (Harmonic, Elliott Wave, etc.)")
    print("‚îú‚îÄ‚îÄ Machine Learning Integration")
    print("‚îú‚îÄ‚îÄ Real-time Market Sentiment Analysis")
    print("‚îú‚îÄ‚îÄ Dynamic Position Sizing (Kelly Criterion)")
    print("‚îú‚îÄ‚îÄ Comprehensive Risk Management")
    print("‚îú‚îÄ‚îÄ Multi-threaded Analysis")
    print("‚îú‚îÄ‚îÄ Advanced Chart Visualization")
    print("‚îî‚îÄ‚îÄ Automated Email Alerts")
    print("=" * 60)

    choice = input("""
üéØ Choose Your Action:
1. üîç Run Single Advanced Scan
2. üîÑ Start Continuous Monitoring
3. üìä Analyze Specific Stock
4. üìà View Historical Performance
5. ‚öôÔ∏è  Configure Settings

Enter choice (1-5): """).strip()

    if choice == '1':
        print("\nüîç Running comprehensive market scan...")
        detector.run_advanced_scanner()

    elif choice == '2':
        interval = input("‚è∞ Enter scan interval in minutes (default 30): ").strip()
        interval = int(interval) if interval.isdigit() else 30
        detector.run_continuous_advanced_scanner(interval)

    elif choice == '3':
        symbol = input("üìä Enter stock symbol (e.g., RELIANCE, TCS): ").strip().upper()
        result = detector.analyze_single_stock(f"{symbol}.NS")

        if result:
            alert_message = detector.send_advanced_alert(result)
            print(f"\n{alert_message}")
        else:
            print(f"‚ùå No significant patterns found for {symbol}")

    elif choice == '4':
        print("üìà Historical performance analysis coming soon...")

    elif choice == '5':
        print("‚öôÔ∏è Configuration management coming soon...")

    print("\n‚úÖ Thank you for using Advanced Indian Stock Pattern Detection System!")
    print("üôè Trade responsibly and manage your risk!")
```

```python


warnings.filterwarnings('ignore')

# Setup logging
logging.basicConfig(
    level=logging.INFO,
    format='%(asctime)s - %(levelname)s - %(message)s',
    handlers=[
        logging.FileHandler('stock_detector.log'),
        logging.StreamHandler()
    ]
)

@dataclass
class TradingSignal:
    symbol: str
    pattern: str
    signal_type: str
    confidence: float
    entry_price: float
    target_price: float
    stop_loss: float
    risk_reward_ratio: float
    volume_confirmation: bool
    sentiment_score: float
    market_regime: str
    timestamp: datetime

class AdvancedStockPatternDetector:
    def __init__(self, config_file='trading_config.json'):
        self.config = self.load_config(config_file)
        self.setup_database()

        # Enhanced Indian stock universe (Top 200 NSE stocks)
        self.nse_stocks = self.load_nse_universe()
        self.bse_stocks = self.load_bse_universe()

        # ML Models
        self.pattern_classifier = None
        self.anomaly_detector = IsolationForest(contamination=0.1, random_state=42)
        self.lstm_model = None

        # Market data
        self.market_data = {}
        self.sentiment_data = {}
        self.option_data = {}

        # Risk management
        self.portfolio_exposure = 0.0
        self.max_portfolio_risk = 0.15  # 15% max portfolio risk

        # Real-time features
        self.active_alerts = {}
        self.paper_trades = []

        self.initialize_models()

    def load_config(self, config_file):
        """Load configuration from JSON file"""
        default_config = {
            "email": {
                "smtp_server": "smtp.gmail.com",
                "smtp_port": 587,
                "sender_email": "",
                "sender_password": "",
                "receiver_emails": []
            },
            "trading": {
                "max_positions": 10,
                "position_size": 0.02,  # 2% per position
                "risk_per_trade": 0.01,  # 1% risk per trade
                "min_volume_ratio": 1.5,
                "min_confidence": 75
            },
            "patterns": {
                "min_pattern_length": 20,
                "max_pattern_length": 100,
                "volatility_threshold": 0.02
            }
        }

        try:
            with open(config_file, 'r') as f:
                config = json.load(f)
            return {**default_config, **config}
        except FileNotFoundError:
            with open(config_file, 'w') as f:
                json.dump(default_config, f, indent=2)
            return default_config

    def setup_database(self):
        """Setup SQLite database for storing historical data"""
        self.conn = sqlite3.connect('stock_patterns.db', check_same_thread=False)
        cursor = self.conn.cursor()

        # Create tables
        cursor.execute('''
            CREATE TABLE IF NOT EXISTS patterns (
                id INTEGER PRIMARY KEY AUTOINCREMENT,
                symbol TEXT,
                pattern_type TEXT,
                confidence REAL,
                detected_date TEXT,
                entry_price REAL,
                target_price REAL,
                stop_loss REAL,
                outcome TEXT,
                pnl REAL
            )
        ''')

        cursor.execute('''
            CREATE TABLE IF NOT EXISTS market_data (
                symbol TEXT,
                date TEXT,
                open REAL,
                high REAL,
                low REAL,
                close REAL,
                volume INTEGER,
                PRIMARY KEY (symbol, date)
            )
        ''')

        self.conn.commit()

    def load_nse_universe(self):
        """Load NSE stock universe"""
        try:
            # Top NSE stocks by market cap
            # nse_top_stocks = [
            #     'RELIANCE.NS', 'TCS.NS', 'HDFCBANK.NS', 'INFY.NS', 'HINDUNILVR.NS',
            #     'ICICIBANK.NS', 'SBIN.NS', 'BHARTIARTL.NS', 'LICI.NS', 'ITC.NS',
            #     'LT.NS', 'HCLTECH.NS', 'MARUTI.NS', 'SUNPHARMA.NS', 'TITAN.NS',
            #     'ONGC.NS', 'TATAMOTORS.NS', 'WIPRO.NS', 'ASIANPAINT.NS', 'ADANIENT.NS',
            #     'COALINDIA.NS', 'BAJFINANCE.NS', 'NESTLEIND.NS', 'POWERGRID.NS',
            #     'NTPC.NS', 'TECHM.NS', 'JSWSTEEL.NS', 'HINDALCO.NS', 'INDUSINDBK.NS',
            #     'HEROMOTOCO.NS', 'BRITANNIA.NS', 'BAJAJFINSV.NS', 'CIPLA.NS',
            #     'ULTRACEMCO.NS', 'DRREDDY.NS', 'EICHERMOT.NS', 'GRASIM.NS',
            #     'AXISBANK.NS', 'KOTAKBANK.NS', 'DIVISLAB.NS', 'TATACONSUM.NS',
            #     'BAJAJ-AUTO.NS', 'APOLLOHOSP.NS', 'ADANIPORTS.NS', 'SHREECEM.NS',
            #     'TATASTEEL.NS', 'BPCL.NS', 'IOC.NS', 'PIDILITIND.NS'
            # ]
            nse_top_stocks = [
            # 'RELIANCE.NS', 'TCS.NS', 'HDFCBANK.NS', 'BHARTIARTL.NS', 'ICICIBANK.NS',
            # 'INFY.NS', 'SBIN.NS', 'LICI.NS', 'HINDUNILVR.NS', 'ITC.NS',
            # 'LT.NS', 'HCLTECH.NS', 'BAJFINANCE.NS', 'ONGC.NS', 'MARUTI.NS',
            # 'SUNPHARMA.NS', 'AXISBANK.NS', 'NTPC.NS', 'TATAMOTORS.NS', 'M&M.NS',
            # 'KOTAKBANK.NS', 'ADANIENT.NS', 'AVENUSUPER.NS', 'ULTRACEMCO.NS', 'HAL.NS',
            # 'ADANIPORTS.NS', 'POWERGRID.NS', 'TITAN.NS', 'COALINDIA.NS', 'ASIANPAINT.NS',
            # 'ADANIGREEN.NS', 'HINDZINC.NS', 'ADANIPOWER.NS', 'WIPRO.NS', 'BAJAJ-AUTO.NS',
            # 'BAJAJFINSV.NS', 'IRFC.NS', 'NESTLEIND.NS', 'SIEMENS.NS', 'IOC.NS',
            # 'BEL.NS', 'JSWSTEEL.NS', 'JIOFIN.NS', 'VBL.NS', 'DLF.NS',
            # 'TATASTEEL.NS', 'ZOMATO.NS', 'TRENT.NS', 'GRASIM.NS', 'PFC.NS',
            # 'VEDL.NS', 'AMBUJACEM.NS', 'LTIM.NS', 'SBILIFE.NS', 'ABB.NS',
            # 'RECLTD.NS', 'PIDILITIND.NS', 'TECHM.NS', 'GAIL.NS', 'HINDALCO.NS',
            # 'HDFCLIFE.NS', 'BRITANNIA.NS', 'LODHA.NS', 'BPCL.NS', 'EICHERMOT.NS',
            'TATAPOWER.NS', 'BANKBARODA.NS', 'MOTHERSON.NS', 'PNB.NS', 'IOB.NS',
            'RVNL.NS', 'CIPLA.NS', 'DIVISLAB.NS', 'JSWENERGY.NS', 'ZYDUSLIFE.NS',
            'TATACONSUM.NS', 'CHOLAFIN.NS', 'TVSMOTOR.NS', 'ADANITRANS.NS', 'INDUSINDBK.NS',
            'DRREDDY.NS', 'DABUR.NS', 'HAVELLS.NS', 'HEROMOTOCO.NS', 'BHEL.NS',
            'CGPOWER.NS', 'TORNTPHARM.NS', 'BAJAJHLDNG.NS', 'IDBI.NS', 'NHPC.NS',
            'UNIONBANK.NS', 'SHRIRAMFIN.NS', 'CANBK.NS', 'MAZAGON.NS', 'BOSCHLTD.NS',
            'UBL.NS', 'ICICIPRULI.NS', 'SHREECEM.NS', 'JINDALSTEL.NS', 'CUMMINSIND.NS',
            'ATGL.NS', 'OFSS.NS', 'SOLARINDS.NS', 'POLYCAB.NS', 'ICICIGI.NS',
            'APOLLOHOSP.NS', 'MAXHEALTH.NS', 'INDHOTEL.NS', 'NAUKRI.NS', 'OIL.NS',
            'GODREJPROP.NS', 'HDFCAMC.NS', 'COLPAL.NS', 'MARICO.NS', 'MANKIND.NS',
            'LUPIN.NS', 'SUZLON.NS', 'AUROPHARMA.NS', 'YESBANK.NS', 'IRCTC.NS',
            'INDIANB.NS', 'TIINDIA.NS', 'PRESTIGE.NS', 'HINDPETRO.NS', 'PERSISTENT.NS',
            'BHARATFORG.NS', 'TORNTPOWER.NS', 'IREDA.NS', 'JSWINFRA.NS', 'SBICARD.NS',
            'SRF.NS', 'MUTHOOTFIN.NS', 'PAYTM.NS', 'ASHOKLEY.NS', 'NMDC.NS',
            'SUPREMEIND.NS', 'COCHINSHIP.NS', 'LINDE.NS', 'UCOBANK.NS', 'FACT.NS',
            'GICRE.NS', 'DIXONTECH.NS', 'SCHAEFFLER.NS', 'PHOENIXLTD.NS', 'OBEROIRLTY.NS',
            'HUDCO.NS', 'CONCOR.NS', 'ALKEM.NS', 'JINDAL.NS', 'BERGEPAINT.NS',
            'PIIND.NS', 'SAIL.NS', 'BALKRISIND.NS', 'KALYANKJIL.NS', 'PATANJALI.NS',
            'ASTRAL.NS', 'ABBOTINDIA.NS', 'THERMAX.NS', 'UNOMINDA.NS', 'MRF.NS',
            'SJVN.NS', 'ABCAPITAL.NS', 'PGHH.NS', 'BANKINDIA.NS', 'CENTRALBANK.NS',
            'LTTS.NS', 'MPHASIS.NS', 'IDFCFIRSTB.NS', 'NYKAA.NS', 'TATACOMM.NS',
            'PETRONET.NS', 'BDL.NS', 'HDFCBANK.NS', 'KPITTECH.NS', 'SUNDARMFIN.NS',
            'VOLTAS.NS', 'FEDERALBNK.NS', 'ACC.NS', 'AUBANK.NS', 'COROMANDEL.NS',
            'HONAUT.NS', 'MAHABANK.NS', 'EXIDEIND.NS', 'ESCORTS.NS', 'PAGEIND.NS',
            'GSK.NS', 'GUJGASLTD.NS', 'PSB.NS', 'LTFH.NS', '3MINDIA.NS',
            'TATAELXSI.NS', 'LICHSGFIN.NS', 'NIACL.NS', 'ADANIWILMAR.NS', 'COFORGE.NS',
            'BIOCON.NS', 'APLAPOLLO.NS', 'AIAENG.NS', 'IRCON.NS', 'TATACHEM.NS',
            'SBILIFE.NS', 'DEEPAKNTR.NS', 'NLCINDIA.NS', 'JUBLFOOD.NS', 'IGL.NS',
            'MRPL.NS', 'MAXFINSERV.NS', 'KEIIND.NS', 'GETINGE.NS', '360ONE.NS',
            'LLOYDSME.NS', 'MMFIN.NS', 'FORTIS.NS', 'SHLTEAUTO.NS', 'ENDURANCE.NS',
            'METROBRAND.NS', 'EMAMILTD.NS', 'GUJFLUORO.NS', 'EMBASSY.NS', 'JKCEMENT.NS',
            'NATIONALUM.NS', 'APOLLOTYRE.NS', 'BLUESTARCO.NS', 'MOTILALOFS.NS', 'GLAND.NS',
            'DALBHARAT.NS', 'BSE.NS', 'CARBORUNIV.NS', 'NBCC.NS', 'TATAINVEST.NS',
            'APARINDS.NS', 'MOTHERSUMI.NS', 'GLOBALHLT.NS', 'SUNTV.NS', 'IPCALAB.NS',
            'CRISIL.NS', 'AEGISCHEM.NS', 'BANDHANBNK.NS', 'HINCOPPER.NS', 'AJANTPHARM.NS',
            'SYNGENE.NS', 'TIMKEN.NS', 'GODREJIND.NS', 'BAYERCROP.NS', 'SUNDRMFAST.NS',
            'POONAWALLA.NS', 'ZFCVINDIA.NS', 'JBCHEPHARM.NS', 'KPRMILL.NS', 'AMARAJABAT.NS',
            'GRINDWELL.NS', 'CROMPTON.NS', 'CHOLAHLDNG.NS', 'SKFINDIA.NS', 'BRIGADE.NS',
            'IRCON.NS', 'JUPITERWAG.NS', 'GRSE.NS', 'TATACHEM.NS', 'HATSUN.NS',
            'CASTROLIND.NS', 'KAYNES.NS', 'MANYAVAR.NS', 'EIHOTEL.NS', 'TVSHLTD.NS',
            'JYOTICNC.NS', 'NHPC.NS', 'GILLETTE.NS', 'CENTURYTEX.NS', 'BASF.NS',
            'WHIRLPOOL.NS', 'DRLALPATH.NS', 'RATNAMANI.NS', 'ISEC.NS', 'AARTI.NS',
            'SUMICHEM.NS', 'CDSL.NS', 'PFIZER.NS', 'NATCOPHARM.NS', 'JBMA.NS',
            'FINOLEX.NS', 'LAURUSLABS.NS', 'KAJARIA.NS', 'ALEMBICLTD.NS', 'RADICO.NS',
            'KECL.NS', 'TEJASNET.NS', 'CIEAUTOM.NS', 'SUVENPHAR.NS', 'CESC.NS',
            'NEROLAC.NS', 'TITAGARH.NS', 'PIRAMALENT.NS', 'SWANENER.NS', 'ELGIEQUIP.NS',
            'IFCI.NS', 'FIVESTAR.NS', 'GODFRYPHLP.NS', 'KALPATPOWR.NS', 'ATUL.NS',
            'RELAXO.NS', 'NCCLTD.NS', 'CELLO.NS', 'PIRAMALENT.NS', 'CYIENT.NS',
            'SIGNATURE.NS', 'MULTI.NS', 'CREDITACC.NS', 'SONATSOFTW.NS', 'BATAINDIA.NS',
            'VINATIORGA.NS', 'VGUARD.NS', 'ADITYA.NS', 'HATSUN.NS', 'PNBHOUSING.NS',
            'CAMS.NS', 'FINPIPE.NS', 'PTCIL.NS', 'BIRLASOFT.NS', 'CHAMBLFERT.NS',
            'IIFL.NS', 'RRKABEL.NS', 'AFFLE.NS', 'ANGELONE.NS', 'WARREERENEW.NS',
            'POLYMED.NS', 'TRITURBINE.NS', 'SHYAMMETL.NS', 'BEML.NS', 'TRIDENT.NS',
            'GESHIP.NS', 'TECHNOE.NS', 'BLUEDART.NS', 'CHALET.NS', 'JYOTHYLAB.NS',
            'NUVAMA.NS', 'RAMCOCEM.NS', 'SOBHA.NS', 'GSPL.NS', 'BIKAJI.NS',
            'INDIAMART.NS', 'IDFC.NS', 'CONCORDBIO.NS', 'SCHNEIDER.NS', 'HBLPOWER.NS',
            'MANAPPURAM.NS', 'JINDALSAW.NS', 'KIRLOSENG.NS', 'ANANTRAJ.NS', 'ASTRAZEN.NS',
            'KIRLOSKAR.NS', 'WELCORP.NS', 'NAVINFLUOR.NS', 'FINCABLES.NS', 'CAPRIGLOBAL.NS',
            'ZENSARTECH.NS', 'AUTHUM.NS', 'KSBPUMPS.NS', 'MAHLOG.NS', 'DATAPATTERN.NS',
            'KARURVYSYA.NS', 'HFCL.NS', 'WELSPUNLIV.NS', 'KRISHNA.NS', 'ACTIONCON.NS',
            'ASAHIINDIA.NS', 'LAKMACH.NS', 'SUPREMEPET.NS', 'RITES.NS', 'GODREJAGRO.NS',
            'REDINGTON.NS', 'GRINFRA.NS', 'RAILTEL.NS', 'FINEORG.NS', 'RKFORGE.NS',
            'ASTERDM.NS', 'DCMSHRIRAM.NS', 'APTUS.NS', 'ANANDRATHI.NS', 'CLEAN.NS',
            'INDIANENGI.NS', 'CENTURYPLY.NS', 'VARDHMANTEXT.NS', 'BBTC.NS', 'MMTC.NS',
            'NEWGEN.NS', 'CPCL.NS', 'HONASA.NS', 'SANOFI.NS', 'DOMS.NS',
            'ZYDUSWELL.NS', 'JAIBALAJI.NS', 'AMBER.NS', 'RBLBANK.NS', 'GODAWARI.NS',
            'ENGINERSIN.NS', 'ERIS.NS', 'OLECTRA.NS', 'BLSINTER.NS', 'AAVAS.NS',
            'EIDPARRY.NS', 'ELECON.NS', 'NETWEB.NS', 'INTELLECT.NS', 'SCI.NS',
            'INGERRAND.NS', 'WESTLIFE.NS', 'CEINFO.NS', 'RAYMOND.NS', 'PNCINFRA.NS',
            'ZEEL.NS', 'PRAJIND.NS', 'GMDC.NS', 'GRANULES.NS', 'UTIAMC.NS',
            'JPPOWER.NS', 'AKZOINDIA.NS', 'TTKPRESTIG.NS', 'TANLA.NS', 'CRAFTSMAN.NS',
            'KFINTECH.NS', 'HAPPSTMNDS.NS', 'ELECTCAST.NS', 'NUVOCO.NS', 'AETHER.NS',
            'VOLTAMP.NS', 'ZENTEC.NS', 'RHIM.NS', 'NAVA.NS', 'RCF.NS',
            'ECLERX.NS', 'CITYUNION.NS', 'INOXINDIA.NS', 'JKBANK.NS', 'BIRLACORPN.NS',
            'THOMASCOOK.NS', 'GRAVITA.NS', 'LEMONTREE.NS', 'TAINWALCHM.NS', 'CERA.NS',
            'CAPLPOINT.NS', 'JKTYRE.NS', 'HAPPYFORGE.NS', 'MINDACORP.NS', 'JUBILANT.NS',
            'TDPOWERSYS.NS', 'TRIL.NS', 'BAJAJELEC.NS', 'CANFINHOME.NS', 'KPIGREEN.NS',
            'RAINBOW.NS', 'ROUTE.NS', 'SHEELAFOAM.NS', 'USHAMART.NS', 'PCBL.NS',
            'INDGRID.NS', 'GENUSPOWER.NS', 'LATENTVIEW.NS', 'KIRIINDUS.NS', 'VALORESTATE.NS',
            'FORCEMOT.NS', 'NEULANDLAB.NS', 'HGINFRA.NS', 'CEATLTD.NS', 'JUSTDIAL.NS',
            'DEEPAKFERT.NS', 'MAHSCOOTER.NS', 'METROPOLIS.NS', 'GPPL.NS', 'VESUVIUS.NS',
            'KNRCON.NS', 'RTNINDIA.NS', 'GLENMARK.NS', 'PGEL.NS', 'TEXRAIL.NS',
            'SAREGAMA.NS', 'SAFARI.NS', 'SAPPHIRE.NS', 'ALKYLAMINE.NS', 'ARVIND.NS',
            'ISGEC.NS', 'GRAPHITE.NS', 'POWERMECH.NS', 'QUESS.NS', 'EQUITAS.NS',
            'BENSHASSL.NS', 'LTFOODS.NS', 'MOIL.NS', 'REDTAPE.NS', 'IONEXCHANGE.NS',
            'AVANTIFEED.NS', 'JKLAKSHMI.NS', 'GALAXY.NS', 'PURVA.NS', 'LLOYDSENGG.NS',
            'VARROC.NS', 'ESABINDIA.NS', 'GNFC.NS', 'MAHLIFE.NS', 'SARDAEN.NS',
            'RATEGAIN.NS', 'JKPAPER.NS', 'JMFINANCIL.NS', 'AZAD.NS', 'JUNIPER.NS',
            'GSFC.NS', 'MAHINDRA.NS', 'RAJESHEXPO.NS', 'EUREKAFORB.NS', 'NATIONALSTD.NS',
            'JUBLINDS.NS', 'MARKSANS.NS', 'CAMPUS.NS', 'HOMEFIRST.NS', 'AHLUWALIA.NS',
            'MIDHANI.NS', 'STRIDES.NS', 'ARCHIES.NS', 'BALRAMCHIN.NS', 'SANDUMA.NS',
            'CMSINFO.NS', 'TRIVENI.NS', 'SHAKTIPUMP.NS', 'KEYSTONE.NS', 'ELANTAS.NS',
            'PRUDENT.NS', 'SBFC.NS', 'ITDCEM.NS', 'KAMA.NS', 'ANURAS.NS',
            'PGHH.NS', 'AURIONPRO.NS', 'SUNTECK.NS', 'STARCEMENT.NS', 'MAHSEAMLES.NS',
            'UJJIVAN.NS', 'KTKBANK.NS', 'SHOPERSTOP.NS', 'ASTRAMICRO.NS', 'KIRLPNU.NS',
            'JUPITERLIFE.NS', 'ELCTRONIC.NS', 'SYRMA.NS', 'MRSBECTORS.NS', 'MASTEK.NS',
            'VTECH.NS', 'SHRIRAMPI.NS', 'INDIABULLS.NS', 'FDC.NS', 'HEG.NS',
            'RELIGARE.NS', 'MEDPLUS.NS', 'VIJAYA.NS', 'INFIBEAM.NS', 'CCL.NS',
            'GANESHHOUC.NS', 'ETHOSLTD.NS', 'GALLANTT.NS', 'INDIAHOUSING.NS', 'DHANUKA.NS',
            'INDOCOUNT.NS', 'SYMPHONY.NS', 'JSWHLDNG.NS', 'ASKAUTOLTD.NS', 'TIPSINDLTD.NS',
            'CHOICEIN.NS', 'DILIPBUILD.NS', 'BLUEJET.NS', 'RESPONIND.NS', 'GARFIBRES.NS',
            'EASYTRIP.NS', 'TIMETECHNO.NS', 'SENCO.NS', 'MAGELLANIC.NS', 'TCNSBRANDS.NS',
            'PRINCEPIPE.NS', 'BALAMINES.NS', 'SUPRAJIT.NS', 'JINDALWORLD.NS', 'SANSERA.NS',
            'DODLA.NS', 'EMUDHRA.NS', 'MANINFRA.NS', 'ASHOKA.NS', 'KRBL.NS',
            'TMBANK.NS', 'NATIONALFERT.NS', 'LAXMIORG.NS', 'KENNAMETAL.NS', 'EPLLTD.NS',
            'SHARDAMOT.NS', 'INDIGOPNTS.NS', 'GREENLAM.NS', 'WELENT.NS', 'PROTEAN.NS',
            'SOUTHBANK.NS', 'GOKEX.NS', 'GABRIEL.NS', 'PDSL.NS', 'NAZARA.NS',
            'BBOX.NS', 'ITDC.NS', 'IFBIND.NS', 'ORCHPHARMA.NS', 'PARADEEP.NS',
            'EPIGRAL.NS', 'HINDFOOD.NS', 'JANASF.NS', 'PICCADILY.NS', 'VIPIND.NS',
            'JAICORPLTD.NS', 'ORIENTCEM.NS', 'RALLIS.NS', 'INSOLATION.NS', 'SHILPAMED.NS',
            'SURYAROSHA.NS', 'ROLEXRINGS.NS', 'PTCIL.NS', 'GMMFAUD.NS', 'ARVIND.NS',
            'MSTCLTD.NS', 'TECHNO.NS', 'NESCO.NS', 'NIITLTD.NS', 'GUJAMBUJA.NS',
            'SISINDIA.NS', 'JKUMAR.NS', 'SHAREINDIA.NS', 'ADITYAVISION.NS', 'DBCORP.NS',
            'PAISALO.NS', 'CSBBANK.NS', 'TDPOWERSYS.NS', 'ALLCARGO.NS', 'CYIENTDLM.NS',
            'EDELWEISS.NS', 'SUDARSCHEM.NS', 'PRIVISPEC.NS', 'BONDADA.NS', 'VSTIND.NS',
            'GOFASHION.NS', 'SUNDARMHLD.NS', 'PRICOL.NS', 'ORIENTELEC.NS', 'KIRLOSIND.NS',
            'IIFLS.NS', 'LUXIND.NS', 'MTARTECH.NS', 'GULFOIL.NS', 'RSYSTEMS.NS',
            'HERITGFOOD.NS', 'ICRA.NS', 'PILANISEC.NS', 'TEAMLEASE.NS', 'UTKARSHSF.NS',
            'GATEWAY.NS', 'KSCL.NS', 'GARWARE.NS', 'MOSCHIP.NS', 'EXICOM.NS',
            'AMIORG.NS', 'THANGAMAYIL.NS', 'JTEKTINDIA.NS', 'VAIBHAVGBL.NS', 'HEIDELBERG.NS',
            'MASFINANCE.NS', 'AARTIPHARMA.NS', 'GHCL.NS', 'AGIGREEN.NS', 'RAMKYINFRA.NS',
            'WONDERLA.NS', 'PATELENG.NS', 'BHARATBIJ.NS', 'JAMNAAUTO.NS', 'PARAS.NS',
            'SPANDANA.NS', 'ENTERO.NS', 'SHANTIGEAR.NS', 'VRLLOG.NS', 'JAIHIND.NS',
            'BANCOINDIA.NS', 'NOCIL.NS', 'SHARDACROP.NS', 'HEALTHGUARD.NS', 'HARSHA.NS',
            'AARTIDRUGS.NS', 'INOXGREEN.NS', 'AVANTEL.NS', 'BALMLAWRIE.NS', 'DYNAMATIC.NS',
            'SANGHIMOTOR.NS', 'KOVAI.NS', 'STYRENIX.NS', 'TILAKNAGAR.NS', 'HAWKINS.NS',
            'TCIEXP.NS', 'NETWEB.NS', 'SHIPPINGCORP.NS', 'LLOYDSENGG.NS', 'FEDFINA.NS',
            'SHILCHAR.NS', 'JAYASWALNECO.NS', 'ROSSARI.NS', 'WPIL.NS', 'SUBROS.NS',
            'HIKAL.NS', 'BOROSIL.NS', 'VENUSPIPES.NS', 'SUNFLAG.NS', 'BHARATRAS.NS',
            'LGBBF.NS', 'EMSWORTH.NS', 'FUSIONMICRO.NS', 'DCXSYSTEMS.NS', 'ISMT.NS',
            'BHAGIRADHA.NS', 'ADVENZYMES.NS', 'WEIZMANN.NS', 'WESTCOAST.NS', 'KDDL.NS',
            'NEOGEN.NS', 'DDEVEVD.NS', 'PITTIENGG.NS', 'SGMART.NS', 'CARTRADE.NS',
            'BALUFORGE.NS', 'HONDAPOWER.NS', 'DCBBANK.NS', 'SULA.NS', 'ASHIANA.NS',
            'KEWALSCHOTT.NS', 'SAVITA.NS', 'NUCLEUS.NS', 'GOPALSNACKS.NS', 'MUTHOOTMF.NS',
            'TIDEWATER.NS', 'FINEOTEX.NS', 'ORIANA.NS', 'IMAGICAAENT.NS', 'THYROCARE.NS',
            'HATHWAY.NS', 'MEDASSIST.NS', 'PEARLPOLY.NS', 'SHAILY.NS', 'APEEJAY.NS',
            'UFLEX.NS', 'GANESHA.NS', 'MANORAMA.NS', 'BAJAJCON.NS', 'GREENPLY.NS',
            'SKIPPER.NS', 'SEAMEC.NS', 'LUMAXAUTO.NS', 'GREENPANEL.NS', 'GRAUER.NS',
            'INDIANMETAL.NS', 'BANARIAMM.NS', 'HINDUJA.NS', 'ZAGGLE.NS', 'SHIVBIMETAL.NS',
            'NIRLON.NS', 'JTLIND.NS', 'HPLELEC.NS', 'YATHARTH.NS', 'SPRIGHT.NS',
            'DATAMATICS.NS', 'OPTIEMUS.NS', 'PTCFINANCE.NS', 'GREAVES.NS', 'ALEMBIC.NS',
            'GENSOL.NS', 'GIPSU.NS', 'CIGNITI.NS', 'BHANSALI.NS', 'SWARAJENG.NS',
            'ANUP.NS', 'APOLLO.NS', 'STEELSTRI.NS', 'GUJTHEM.NS', 'DELTACORP.NS',
            'NAVNETEDUL.NS', 'DREDGING.NS', 'GUFICBIO.NS', 'INNOVCAP.NS', 'SANDHAR.NS',
            'VSTTILLERS.NS', 'FIEMIND.NS', 'MPSLTD.NS', 'KCPSUG.NS', 'PREMEXPLN.NS',
            'JINDALPOLY.NS', 'LAOPALA.NS', 'THEJO.NS', 'AVALON.NS', 'MARINELEC.NS',
            'IDEA.NS', 'PRAKASH.NS', 'INDOSTAR.NS', 'ASHAPURA.NS', 'INDIAGLYC.NS',
            'MAITHANALL.NS', 'VINDHYA.NS', 'HINDOILEXP.NS', 'REPCOHOME.NS', 'RPGLIFE.NS',
            'POLYPLEX.NS', 'QUICKHEAL.NS', 'STYLAMIND.NS', 'TVSSRICH.NS', 'ARVINDSMART.NS',
            'FLAIRW.NS', 'DALMIA.NS', 'HINDWARE.NS', 'ARTEMISMED.NS', 'NRBEARING.NS',
            'UJAAS.NS', 'SOMANYCERA.NS', 'TINNARUBBR.NS', 'MARATHON.NS', 'VADILAL.NS',
            'SPECTRUM.NS', 'AUTOAXLES.NS', 'SAKSOFT.NS', 'SUPRIYA.NS', 'KFATECH.NS',
            'UNIVCABLES.NS', 'MMFORGNG.NS', 'GOODLUCK.NS', 'SEPC.NS', 'BFUTILITIES.NS',
            'CARERATING.NS', 'VERTOZ.NS', 'DOLIND.NS', 'SALASAR.NS', 'SHALBY.NS',
            'HLEGLAS.NS', 'SMLISUZU.NS', 'RAJRATAN.NS', 'SKYGOLD.NS', 'DPABHUSHAN.NS',
            'EVEREADY.NS', 'MANINDS.NS', 'INDOREM.NS', 'WENDT.NS', 'JASH.NS',
            'REFEX.NS', 'MOREPENLAB.NS', 'NILKAMAL.NS', 'JOHNCOCKRILL.NS', 'FOSECO.NS',
            'GENESYS.NS', 'SINDHUTR.NS', 'EIHASSON.NS', 'VERITAS.NS', 'HUHTAMAKI.NS',
            'PRECWIRE.NS', 'ACCELYA.NS', 'VISHNU.NS', 'V2RETAIL.NS', 'KPENERGY.NS',
            'JEENASIKHO.NS', 'NOVARTIS.NS', 'SERVOTECH.NS', 'RASHI.NS', 'KPGREEN.NS',
            'DISA.NS', 'GOODYEAR.NS', 'VENKEYS.NS', 'VISHNUCHE.NS', 'LUMAXIND.NS',
            'GEOJITFSL.NS', 'PSPPROJECT.NS', 'KALYANIINV.NS', 'SHALKAR.NS'
            ]
            return nse_top_stocks
        except Exception as e:
            logging.error(f"Error loading NSE universe: {e}")
            return []

    def load_bse_universe(self):
        """Load BSE stock universe"""
        return [stock.replace('.NS', '.BO') for stock in self.load_nse_universe()]

    def get_enhanced_stock_data(self, symbol, period='2y'):
        """Enhanced stock data fetching with multiple sources"""
        try:
            # Primary source: yfinance
            stock = yf.Ticker(symbol)
            data = stock.history(period=period)

            if data.empty:
                logging.warning(f"No data from yfinance for {symbol}")
                return None

            # Add enhanced features
            data = self.add_advanced_technical_indicators(data)
            data = self.add_market_microstructure_features(data)

            return data

        except Exception as e:
            logging.error(f"Error fetching enhanced data for {symbol}: {e}")
            return None

    def add_advanced_technical_indicators(self, data):
        """Add comprehensive technical indicators"""
        try:
            # Basic indicators
            data['RSI'] = ta.momentum.RSIIndicator(data['Close'], window=14).rsi()
            data['RSI_2'] = ta.momentum.RSIIndicator(data['Close'], window=2).rsi()
            data['MACD'] = ta.trend.MACD(data['Close']).macd()
            data['MACD_Signal'] = ta.trend.MACD(data['Close']).macd_signal()
            data['MACD_Histogram'] = data['MACD'] - data['MACD_Signal']

            # Multiple timeframe EMAs
            for period in [8, 13, 21, 34, 55, 89, 144]:
                data[f'EMA_{period}'] = ta.trend.EMAIndicator(data['Close'], window=period).ema_indicator()

            # Bollinger Bands with multiple periods
            for period in [20, 50]:
                bb = ta.volatility.BollingerBands(data['Close'], window=period)
                data[f'BB_Upper_{period}'] = bb.bollinger_hband()
                data[f'BB_Lower_{period}'] = bb.bollinger_lband()
                data[f'BB_Width_{period}'] = (data[f'BB_Upper_{period}'] - data[f'BB_Lower_{period}']) / data['Close']
                data[f'BB_Position_{period}'] = (data['Close'] - data[f'BB_Lower_{period}']) / (data[f'BB_Upper_{period}'] - data[f'BB_Lower_{period}'])

            # Volume indicators
            data['Volume_SMA_20'] = data['Volume'].rolling(window=20).mean()
            data['Volume_SMA_50'] = data['Volume'].rolling(window=50).mean()
            data['Volume_Ratio'] = data['Volume'] / data['Volume_SMA_20']
            data['OBV'] = ta.volume.OnBalanceVolumeIndicator(data['Close'], data['Volume']).on_balance_volume()
            data['VWAP'] = (data['Volume'] * (data['High'] + data['Low'] + data['Close']) / 3).cumsum() / data['Volume'].cumsum()

            # Momentum indicators
            data['ADX'] = ta.trend.ADXIndicator(data['High'], data['Low'], data['Close']).adx()
            data['CCI'] = ta.trend.CCIIndicator(data['High'], data['Low'], data['Close']).cci()
            data['Williams_R'] = ta.momentum.WilliamsRIndicator(data['High'], data['Low'], data['Close']).williams_r()
            data['Stoch_K'] = ta.momentum.StochasticOscillator(data['High'], data['Low'], data['Close']).stoch()
            data['Stoch_D'] = ta.momentum.StochasticOscillator(data['High'], data['Low'], data['Close']).stoch_signal()

            # Volatility indicators
            data['ATR'] = ta.volatility.AverageTrueRange(data['High'], data['Low'], data['Close']).average_true_range()
            data['Volatility'] = data['Close'].pct_change().rolling(window=20).std() * np.sqrt(252)

            # Support and Resistance
            data['Resistance'] = data['High'].rolling(window=20).max()
            data['Support'] = data['Low'].rolling(window=20).min()

            # Price patterns
            data['Higher_High'] = (data['High'] > data['High'].shift(1)) & (data['High'].shift(1) > data['High'].shift(2))
            data['Lower_Low'] = (data['Low'] < data['Low'].shift(1)) & (data['Low'].shift(1) < data['Low'].shift(2))

            return data

        except Exception as e:
            logging.error(f"Error adding technical indicators: {e}")
            return data

    def add_market_microstructure_features(self, data):
        """Add market microstructure and behavioral features"""
        try:
            # Price action features
            data['Body_Size'] = abs(data['Close'] - data['Open']) / data['Open']
            data['Upper_Shadow'] = (data['High'] - np.maximum(data['Open'], data['Close'])) / data['Open']
            data['Lower_Shadow'] = (np.minimum(data['Open'], data['Close']) - data['Low']) / data['Open']
            data['Range'] = (data['High'] - data['Low']) / data['Open']

            # Gap analysis
            data['Gap'] = (data['Open'] - data['Close'].shift(1)) / data['Close'].shift(1)
            data['Gap_Up'] = data['Gap'] > 0.02
            data['Gap_Down'] = data['Gap'] < -0.02

            # Momentum features
            for period in [1, 3, 5, 10]:
                data[f'Return_{period}d'] = data['Close'].pct_change(period)
                data[f'Vol_Adj_Return_{period}d'] = data[f'Return_{period}d'] / data['Volatility']

            # Regime detection
            data['Trend_Strength'] = np.where(
                data['Close'] > data['EMA_21'],
                (data['Close'] - data['EMA_21']) / data['EMA_21'],
                (data['Close'] - data['EMA_21']) / data['EMA_21']
            )

            # Volume profile approximation
            data['Volume_Profile'] = data['Volume'] / data['Volume'].rolling(window=50).max()

            return data

        except Exception as e:
            logging.error(f"Error adding microstructure features: {e}")
            return data

    def detect_advanced_patterns(self, data, symbol):
        """Detect advanced chart patterns using ML and statistical methods"""
        patterns = []

        try:
            # Traditional patterns with ML enhancement
            patterns.extend(self.detect_harmonic_patterns(data))
            patterns.extend(self.detect_elliott_wave_patterns(data))
            patterns.extend(self.detect_volume_patterns(data))
            patterns.extend(self.detect_breakout_patterns(data))
            patterns.extend(self.detect_reversal_patterns(data))
            patterns.extend(self.detect_continuation_patterns(data))
            patterns.extend(self.detect_anomaly_patterns(data))

            # ML-based pattern detection
            if self.pattern_classifier:
                ml_patterns = self.detect_ml_patterns(data)
                patterns.extend(ml_patterns)

        except Exception as e:
            logging.error(f"Error detecting patterns for {symbol}: {e}")

        return patterns

    def detect_harmonic_patterns(self, data):
        """Detect harmonic patterns (Gartley, Butterfly, Bat, Crab)"""
        patterns = []

        try:
            if len(data) < 100:
                return patterns

            # Simplified harmonic pattern detection
            closes = data['Close'].values
            highs = data['High'].values
            lows = data['Low'].values

            # Find significant pivots
            high_pivots = find_peaks(highs, distance=10, prominence=np.std(highs)*0.5)[0]
            low_pivots = find_peaks(-lows, distance=10, prominence=np.std(lows)*0.5)[0]

            if len(high_pivots) >= 3 and len(low_pivots) >= 2:
                # Check for potential Gartley pattern
                pattern = self.validate_gartley_pattern(data, high_pivots, low_pivots)
                if pattern:
                    patterns.append(pattern)

        except Exception as e:
            logging.error(f"Error detecting harmonic patterns: {e}")

        return patterns

    def validate_gartley_pattern(self, data, high_pivots, low_pivots):
        """Validate Gartley harmonic pattern"""
        try:
            # Get recent pivots
            if len(high_pivots) < 2 or len(low_pivots) < 2:
                return None

            # Simplified Gartley validation
            current_price = data['Close'].iloc[-1]

            # Check Fibonacci retracements (simplified)
            recent_high = data['High'].iloc[high_pivots[-1]]
            recent_low = data['Low'].iloc[low_pivots[-1]]

            retracement = (current_price - recent_low) / (recent_high - recent_low)

            if 0.618 <= retracement <= 0.786:  # Gartley B point retracement
                return {
                    'pattern': 'Gartley Pattern',
                    'type': 'bullish' if current_price < recent_high else 'bearish',
                    'confidence': 80,
                    'entry_level': current_price,
                    'target': recent_high if current_price < recent_high else recent_low,
                    'retracement_level': retracement
                }

        except Exception as e:
            logging.error(f"Error validating Gartley pattern: {e}")

        return None

    def detect_elliott_wave_patterns(self, data):
        """Detect Elliott Wave patterns"""
        patterns = []

        try:
            if len(data) < 50:
                return patterns

            # Simplified Elliott Wave detection
            closes = data['Close'].values

            # Find wave structure using pivot points
            pivots = self.find_elliott_pivots(closes)

            if len(pivots) >= 5:
                wave_pattern = self.analyze_wave_structure(pivots, closes)
                if wave_pattern:
                    patterns.append(wave_pattern)

        except Exception as e:
            logging.error(f"Error detecting Elliott Wave patterns: {e}")

        return patterns

    def find_elliott_pivots(self, prices):
        """Find pivots for Elliott Wave analysis"""
        high_pivots = find_peaks(prices, distance=5)[0]
        low_pivots = find_peaks(-prices, distance=5)[0]

        # Combine and sort pivots
        all_pivots = list(high_pivots) + list(low_pivots)
        all_pivots.sort()

        return all_pivots

    def analyze_wave_structure(self, pivots, prices):
        """Analyze Elliott Wave structure"""
        try:
            if len(pivots) < 5:
                return None

            # Get recent 5 waves
            recent_pivots = pivots[-5:]
            wave_prices = [prices[i] for i in recent_pivots]

            # Simple wave validation (5-wave impulse pattern)
            if len(wave_prices) == 5:
                # Check for alternation and extension
                wave1 = abs(wave_prices[1] - wave_prices[0])
                wave3 = abs(wave_prices[3] - wave_prices[2])
                wave5 = abs(wave_prices[4] - wave_prices[3])

                # Wave 3 should be longest (simplified rule)
                if wave3 > wave1 and wave3 > wave5:
                    return {
                        'pattern': 'Elliott Wave Impulse',
                        'type': 'trending',
                        'confidence': 70,
                        'wave_count': 5,
                        'current_wave': 5
                    }

        except Exception as e:
            logging.error(f"Error analyzing wave structure: {e}")

        return None

    def detect_volume_patterns(self, data):
        """Detect volume-based patterns"""
        patterns = []

        try:
            # Volume breakout pattern
            avg_volume = data['Volume'].rolling(window=20).mean()
            volume_spike = data['Volume'] > (avg_volume * 2)
            price_breakout = data['Close'] > data['Close'].rolling(window=20).max().shift(1)

            if volume_spike.iloc[-1] and price_breakout.iloc[-1]:
                patterns.append({
                    'pattern': 'Volume Breakout',
                    'type': 'bullish',
                    'confidence': 85,
                    'volume_ratio': data['Volume'].iloc[-1] / avg_volume.iloc[-1],
                    'breakout_level': data['Close'].rolling(window=20).max().shift(1).iloc[-1]
                })

            # Volume divergence pattern
            price_trend = data['Close'].rolling(window=10).apply(lambda x: stats.linregress(range(len(x)), x)[0])
            volume_trend = data['Volume'].rolling(window=10).apply(lambda x: stats.linregress(range(len(x)), x)[0])

            if (price_trend.iloc[-1] > 0 and volume_trend.iloc[-1] < 0) or \
               (price_trend.iloc[-1] < 0 and volume_trend.iloc[-1] > 0):
                patterns.append({
                    'pattern': 'Volume Divergence',
                    'type': 'warning',
                    'confidence': 75,
                    'price_trend': price_trend.iloc[-1],
                    'volume_trend': volume_trend.iloc[-1]
                })

        except Exception as e:
            logging.error(f"Error detecting volume patterns: {e}")

        return patterns

    def detect_breakout_patterns(self, data):
        """Detect breakout patterns with confirmation"""
        patterns = []

        try:
            # Range breakout
            range_high = data['High'].rolling(window=20).max()
            range_low = data['Low'].rolling(window=20).min()
            range_size = (range_high - range_low) / data['Close']

            # Look for consolidation followed by breakout
            consolidation = range_size < 0.15  # Less than 15% range
            breakout_up = data['Close'] > range_high.shift(1)
            breakout_down = data['Close'] < range_low.shift(1)

            if consolidation.rolling(window=10).sum().iloc[-1] >= 7:  # 7 out of 10 days consolidating
                if breakout_up.iloc[-1]:
                    patterns.append({
                        'pattern': 'Range Breakout (Bullish)',
                        'type': 'bullish',
                        'confidence': 80,
                        'breakout_level': range_high.iloc[-2],
                        'target': range_high.iloc[-2] + (range_high.iloc[-2] - range_low.iloc[-2]) * 0.618
                    })
                elif breakout_down.iloc[-1]:
                    patterns.append({
                        'pattern': 'Range Breakout (Bearish)',
                        'type': 'bearish',
                        'confidence': 80,
                        'breakout_level': range_low.iloc[-2],
                        'target': range_low.iloc[-2] - (range_high.iloc[-2] - range_low.iloc[-2]) * 0.618
                    })

        except Exception as e:
            logging.error(f"Error detecting breakout patterns: {e}")

        return patterns

    def detect_reversal_patterns(self, data):
        """Detect reversal patterns"""
        patterns = []

        try:
            # Hammer/Doji patterns
            for i in range(-5, 0):  # Check last 5 candles
                if abs(i) > len(data):
                    continue

                candle = data.iloc[i]
                body_size = abs(candle['Close'] - candle['Open']) / candle['Open']
                upper_shadow = (candle['High'] - max(candle['Open'], candle['Close'])) / candle['Open']
                lower_shadow = (min(candle['Open'], candle['Close']) - candle['Low']) / candle['Open']

                # Hammer pattern
                if (body_size < 0.01 and lower_shadow > 0.02 and upper_shadow < 0.005):
                    patterns.append({
                        'pattern': 'Hammer',
                        'type': 'bullish',
                        'confidence': 70,
                        'candle_index': i,
                        'body_size': body_size,
                        'lower_shadow': lower_shadow
                    })

                # Shooting star pattern
                if (body_size < 0.01 and upper_shadow > 0.02 and lower_shadow < 0.005):
                    patterns.append({
                        'pattern': 'Shooting Star',
                        'type': 'bearish',
                        'confidence': 70,
                        'candle_index': i,
                        'body_size': body_size,
                        'upper_shadow': upper_shadow
                    })

        except Exception as e:
            logging.error(f"Error detecting reversal patterns: {e}")

        return patterns

    def detect_continuation_patterns(self, data):
        """Detect continuation patterns"""
        patterns = []

        try:
            # Flag pattern
            if len(data) >= 30:
                # Look for strong move followed by consolidation
                price_change_20 = (data['Close'].iloc[-20] - data['Close'].iloc[-30]) / data['Close'].iloc[-30]
                recent_volatility = data['Close'].iloc[-10:].pct_change().std()

                if abs(price_change_20) > 0.1 and recent_volatility < 0.02:  # 10% move, then low volatility
                    pattern_type = 'bullish' if price_change_20 > 0 else 'bearish'
                    patterns.append({
                        'pattern': f'Flag Pattern ({pattern_type.title()})',
                        'type': pattern_type,
                        'confidence': 75,
                        'pole_size': abs(price_change_20),
                        'consolidation_volatility': recent_volatility
                    })

        except Exception as e:
            logging.error(f"Error detecting continuation patterns: {e}")

        return patterns

    def detect_anomaly_patterns(self, data):
        """Detect anomalous patterns using statistical methods"""
        patterns = []

        try:
            # Prepare features for anomaly detection
            features = []
            for col in ['Close', 'Volume', 'RSI', 'MACD']:
                if col in data.columns:
                    features.append(data[col].fillna(method='ffill').values[-50:])  # Last 50 days

            if features:
                features_array = np.column_stack(features)

                # Detect anomalies
                anomalies = self.anomaly_detector.fit_predict(features_array)

                # If recent data points are anomalies
                if anomalies[-1] == -1:  # -1 indicates anomaly
                    patterns.append({
                        'pattern': 'Statistical Anomaly',
                        'type': 'warning',
                        'confidence': 60,
                        'anomaly_score': self.anomaly_detector.score_samples(features_array)[-1]
                    })

        except Exception as e:
            logging.error(f"Error detecting anomaly patterns: {e}")

        return patterns

    def detect_ml_patterns(self, data):
        """Use ML model to detect patterns"""
        patterns = []

        try:
            if self.pattern_classifier and len(data) >= 50:
                # Prepare features
                features = self.prepare_ml_features(data)

                if features is not None:
                    # Predict pattern
                    prediction = self.pattern_classifier.predict([features])
                    confidence = self.pattern_classifier.predict_proba([features]).max()

                    if confidence > 0.7:  # High confidence threshold
                        pattern_names = ['No Pattern', 'Bullish', 'Bearish', 'Neutral']
                        patterns.append({
                            'pattern': f'ML Pattern: {pattern_names[prediction[0]]}',
                            'type': pattern_names[prediction[0]].lower(),
                            'confidence': confidence * 100,
                            'ml_prediction': True
                        })

        except Exception as e:
            logging.error(f"Error with ML pattern detection: {e}")

        return patterns

    def prepare_ml_features(self, data):
        """Prepare features for ML model"""
        try:
            features = []

            # Recent price action features
            features.extend([
                data['RSI'].iloc[-1] if 'RSI' in data.columns else 50,
                data['MACD'].iloc[-1] if 'MACD' in data.columns else 0,
                data['Volume_Ratio'].iloc[-1] if 'Volume_Ratio' in data.columns else 1,
                data['ATR'].iloc[-1] if 'ATR' in data.columns else 0.02,
                data['BB_Position_20'].iloc[-1] if 'BB_Position_20' in data.columns else 0.5
            ])

            # Price momentum features
            if len(data) >= 10:
                features.extend([
                    data['Close'].pct_change(5).iloc[-1],
                    data['Close'].pct_change(10).iloc[-1],
                    data['Volume'].rolling(5).mean().iloc[-1] / data['Volume'].rolling(20).mean().iloc[-1]
                ])

            return np.array(features) if len(features) > 0 else None

        except Exception as e:
            logging.error(f"Error preparing ML features: {e}")
            return None

    def initialize_models(self):
        """Initialize ML models"""
        try:
            # Initialize pattern classifier with dummy data
            X_dummy = np.random.rand(100, 8)  # 8 features
            y_dummy = np.random.randint(0, 4, 100)  # 4 classes

            self.pattern_classifier = RandomForestClassifier(
                n_estimators=100,
                max_depth=10,
                random_state=42
            )
            self.pattern_classifier.fit(X_dummy, y_dummy)

            logging.info("ML models initialized successfully")

        except Exception as e:
            logging.error(f"Error initializing ML models: {e}")

    def calculate_position_size(self, signal, account_balance=100000):
        """Calculate optimal position size using Kelly Criterion"""
        try:
            # Kelly Criterion: f = (bp - q) / b
            # where f = fraction of capital to wager
            # b = odds received on the wager (target/risk)
            # p = probability of winning
            # q = probability of losing (1 - p)

            win_prob = signal.confidence / 100
            loss_prob = 1 - win_prob

            if signal.target_price and signal.stop_loss:
                # Calculate risk-reward ratio
                risk = abs(signal.entry_price - signal.stop_loss) / signal.entry_price
                reward = abs(signal.target_price - signal.entry_price) / signal.entry_price

                if risk > 0:
                    odds = reward / risk
                    kelly_fraction = (win_prob * odds - loss_prob) / odds

                    # Apply Kelly with safety factor (25% of Kelly)
                    kelly_fraction = max(0, min(kelly_fraction * 0.25, 0.05))  # Max 5% per trade

                    position_size = account_balance * kelly_fraction

                    return {
                        'position_size': position_size,
                        'kelly_fraction': kelly_fraction,
                        'risk_reward_ratio': odds,
                        'risk_amount': position_size * risk
                    }

            # Default position sizing (2% of account)
            return {
                'position_size': account_balance * 0.02,
                'kelly_fraction': 0.02,
                'risk_reward_ratio': 1.0,
                'risk_amount': account_balance * 0.02 * 0.02
            }

        except Exception as e:
            logging.error(f"Error calculating position size: {e}")
            return {'position_size': account_balance * 0.01, 'kelly_fraction': 0.01}

    def get_market_sentiment(self, symbol):
        """Get market sentiment from multiple sources"""
        try:
            sentiment_score = 0.0

            # VIX-based market fear gauge (simplified)
            try:
                vix = yf.Ticker("^VIX")
                vix_data = vix.history(period="5d")
                if not vix_data.empty:
                    current_vix = vix_data['Close'].iloc[-1]
                    # Normalize VIX to sentiment score (-1 to 1)
                    sentiment_score += (25 - current_vix) / 25  # Inverted VIX
            except:
                pass

            # Dollar Index impact
            try:
                dxy = yf.Ticker("DX-Y.NYB")
                dxy_data = dxy.history(period="5d")
                if not dxy_data.empty:
                    dxy_change = dxy_data['Close'].pct_change(5).iloc[-1]
                    sentiment_score -= dxy_change * 2  # Strong dollar = negative for Indian stocks
            except:
                pass

            # Crude oil impact (India is oil importer)
            try:
                oil = yf.Ticker("CL=F")
                oil_data = oil.history(period="5d")
                if not oil_data.empty:
                    oil_change = oil_data['Close'].pct_change(5).iloc[-1]
                    sentiment_score -= oil_change * 0.5  # Higher oil = negative for India
            except:
                pass

            # Normalize sentiment score
            sentiment_score = max(-1, min(1, sentiment_score))

            return {
                'sentiment_score': sentiment_score,
                'sentiment_label': self.get_sentiment_label(sentiment_score)
            }

        except Exception as e:
            logging.error(f"Error getting market sentiment: {e}")
            return {'sentiment_score': 0.0, 'sentiment_label': 'Neutral'}

    def get_sentiment_label(self, score):
        """Convert sentiment score to label"""
        if score > 0.3:
            return 'Very Bullish'
        elif score > 0.1:
            return 'Bullish'
        elif score > -0.1:
            return 'Neutral'
        elif score > -0.3:
            return 'Bearish'
        else:
            return 'Very Bearish'

    def get_market_regime(self, symbol):
        """Determine current market regime"""
        try:
            # Get NIFTY data for market regime
            nifty = yf.Ticker("^NSEI")
            nifty_data = nifty.history(period="6mo")

            if nifty_data.empty:
                return 'Unknown'

            # Calculate regime indicators
            returns = nifty_data['Close'].pct_change().dropna()
            volatility = returns.rolling(window=20).std() * np.sqrt(252)
            trend = nifty_data['Close'].rolling(window=50).apply(lambda x: stats.linregress(range(len(x)), x)[0])

            current_vol = volatility.iloc[-1]
            current_trend = trend.iloc[-1]

            # Determine regime
            if current_vol > 0.25:  # High volatility
                if abs(current_trend) > 0.1:
                    return 'High Volatility Trending'
                else:
                    return 'High Volatility Ranging'
            else:  # Low volatility
                if current_trend > 0.05:
                    return 'Low Volatility Bull Market'
                elif current_trend < -0.05:
                    return 'Low Volatility Bear Market'
                else:
                    return 'Low Volatility Ranging'

        except Exception as e:
            logging.error(f"Error determining market regime: {e}")
            return 'Unknown'

    def generate_advanced_signal(self, symbol, patterns, data):
        """Generate advanced trading signals with comprehensive analysis"""
        try:
            if not patterns:
                return None

            # Get market context
            sentiment = self.get_market_sentiment(symbol)
            market_regime = self.get_market_regime(symbol)

            # Select best pattern
            best_pattern = max(patterns, key=lambda x: x.get('confidence', 0))

            current_price = data['Close'].iloc[-1]
            atr = data['ATR'].iloc[-1] if 'ATR' in data.columns else current_price * 0.02

            # Generate signal based on pattern type
            signal_type = 'WATCH'
            entry_price = current_price
            target_price = None
            stop_loss = None

            if best_pattern['type'] == 'bullish':
                # Bullish signal logic
                if sentiment['sentiment_score'] > -0.2:  # Not too bearish
                    signal_type = 'BUY'
                    entry_price = current_price

                    # Dynamic target based on ATR and pattern
                    if 'target' in best_pattern:
                        target_price = best_pattern['target']
                    else:
                        target_price = current_price * (1 + 2 * atr / current_price)  # 2 ATR target

                    # Dynamic stop loss
                    stop_loss = current_price * (1 - 1.5 * atr / current_price)  # 1.5 ATR stop

            elif best_pattern['type'] == 'bearish':
                # Bearish signal logic
                if sentiment['sentiment_score'] < 0.2:  # Not too bullish
                    signal_type = 'SELL'
                    entry_price = current_price

                    # Dynamic target based on ATR and pattern
                    if 'target' in best_pattern:
                        target_price = best_pattern['target']
                    else:
                        target_price = current_price * (1 - 2 * atr / current_price)  # 2 ATR target

                    # Dynamic stop loss
                    stop_loss = current_price * (1 + 1.5 * atr / current_price)  # 1.5 ATR stop

            # Calculate risk-reward ratio
            risk_reward_ratio = 0
            if target_price and stop_loss:
                risk = abs(entry_price - stop_loss)
                reward = abs(target_price - entry_price)
                risk_reward_ratio = reward / risk if risk > 0 else 0

            # Volume confirmation
            volume_confirmation = False
            if 'Volume_Ratio' in data.columns:
                volume_confirmation = data['Volume_Ratio'].iloc[-1] > 1.5

            # Create advanced signal
            signal = TradingSignal(
                symbol=symbol.replace('.NS', ''),
                pattern=best_pattern['pattern'],
                signal_type=signal_type,
                confidence=min(best_pattern['confidence'] *
                             (1 + sentiment['sentiment_score'] * 0.1), 100),  # Sentiment adjustment
                entry_price=entry_price,
                target_price=target_price,
                stop_loss=stop_loss,
                risk_reward_ratio=risk_reward_ratio,
                volume_confirmation=volume_confirmation,
                sentiment_score=sentiment['sentiment_score'],
                market_regime=market_regime,
                timestamp=datetime.now()
            )

            return signal

        except Exception as e:
            logging.error(f"Error generating advanced signal for {symbol}: {e}")
            return None

    def scan_stocks_parallel(self, max_workers=5):
        """Scan stocks in parallel for better performance"""
        results = []

        with ThreadPoolExecutor(max_workers=max_workers) as executor:
            # Submit all tasks
            future_to_symbol = {
                executor.submit(self.analyze_single_stock, symbol): symbol
                for symbol in self.nse_stocks[:]  # Top 50 stocks--------------------------------------------------
            }

            # Collect results
            for future in as_completed(future_to_symbol):
                symbol = future_to_symbol[future]
                try:
                    result = future.result()
                    if result:
                        results.append(result)
                        logging.info(f"‚úÖ Analyzed {symbol}")
                except Exception as e:
                    logging.error(f"‚ùå Error analyzing {symbol}: {e}")

        return results

    def analyze_single_stock(self, symbol):
        """Analyze a single stock comprehensively"""
        try:
            # Get enhanced stock data
            data = self.get_enhanced_stock_data(symbol)
            if data is None or len(data) < 50:
                return None

            # Detect patterns
            patterns = self.detect_advanced_patterns(data, symbol)
            if not patterns:
                return None

            # Generate signal
            signal = self.generate_advanced_signal(symbol, patterns, data)
            if not signal:
                return None

            # Calculate position sizing
            position_info = self.calculate_position_size(signal)

            # Store in database
            self.store_pattern_in_db(signal, patterns)

            return {
                'signal': signal,
                'patterns': patterns,
                'position_info': position_info,
                'technical_data': {
                    'rsi': data['RSI'].iloc[-1] if 'RSI' in data.columns else None,
                    'macd': data['MACD'].iloc[-1] if 'MACD' in data.columns else None,
                    'volume_ratio': data['Volume_Ratio'].iloc[-1] if 'Volume_Ratio' in data.columns else None,
                    'atr': data['ATR'].iloc[-1] if 'ATR' in data.columns else None
                }
            }

        except Exception as e:
            logging.error(f"Error in single stock analysis for {symbol}: {e}")
            return None

    def store_pattern_in_db(self, signal, patterns):
        """Store detected patterns in database"""
        try:
            cursor = self.conn.cursor()

            for pattern in patterns:
                cursor.execute('''
                    INSERT INTO patterns (symbol, pattern_type, confidence, detected_date,
                                        entry_price, target_price, stop_loss, outcome, pnl)
                    VALUES (?, ?, ?, ?, ?, ?, ?, ?, ?)
                ''', (
                    signal.symbol,
                    pattern['pattern'],
                    pattern.get('confidence', 0),
                    signal.timestamp.isoformat(),
                    signal.entry_price,
                    signal.target_price,
                    signal.stop_loss,
                    'PENDING',
                    0.0
                ))

            self.conn.commit()

        except Exception as e:
            logging.error(f"Error storing pattern in database: {e}")

    def create_advanced_visualization(self, symbol, data, patterns, signal):
        """Create advanced chart visualization"""
        try:
            fig = make_subplots(
                rows=4, cols=1,
                shared_xaxes=True,
                vertical_spacing=0.03,
                subplot_titles=('Price & Patterns', 'Volume', 'RSI', 'MACD'),
                row_heights=[0.5, 0.2, 0.15, 0.15]
            )

            # Candlestick chart
            fig.add_trace(
                go.Candlestick(
                    x=data.index,
                    open=data['Open'],
                    high=data['High'],
                    low=data['Low'],
                    close=data['Close'],
                    name='Price'
                ),
                row=1, col=1
            )

            # Add EMAs
            if 'EMA_21' in data.columns:
                fig.add_trace(
                    go.Scatter(x=data.index, y=data['EMA_21'], name='EMA 21',
                             line=dict(color='orange', width=1)),
                    row=1, col=1
                )

            if 'EMA_55' in data.columns:
                fig.add_trace(
                    go.Scatter(x=data.index, y=data['EMA_55'], name='EMA 55',
                             line=dict(color='blue', width=1)),
                    row=1, col=1
                )

            # Add signal markers
            if signal.signal_type in ['BUY', 'SELL']:
                color = 'green' if signal.signal_type == 'BUY' else 'red'
                fig.add_trace(
                    go.Scatter(
                        x=[data.index[-1]],
                        y=[signal.entry_price],
                        mode='markers',
                        marker=dict(size=15, color=color, symbol='triangle-up' if signal.signal_type == 'BUY' else 'triangle-down'),
                        name=f'{signal.signal_type} Signal'
                    ),
                    row=1, col=1
                )

            # Volume
            fig.add_trace(
                go.Bar(x=data.index, y=data['Volume'], name='Volume',
                      marker_color='lightblue'),
                row=2, col=1
            )

            # RSI
            if 'RSI' in data.columns:
                fig.add_trace(
                    go.Scatter(x=data.index, y=data['RSI'], name='RSI',
                             line=dict(color='purple')),
                    row=3, col=1
                )
                # RSI levels
                fig.add_hline(y=70, line_dash="dash", line_color="red", row=3, col=1)
                fig.add_hline(y=30, line_dash="dash", line_color="green", row=3, col=1)

            # MACD
            if 'MACD' in data.columns and 'MACD_Signal' in data.columns:
                fig.add_trace(
                    go.Scatter(x=data.index, y=data['MACD'], name='MACD',
                             line=dict(color='blue')),
                    row=4, col=1
                )
                fig.add_trace(
                    go.Scatter(x=data.index, y=data['MACD_Signal'], name='MACD Signal',
                             line=dict(color='red')),
                    row=4, col=1
                )

            fig.update_layout(
                title=f'{symbol} - {signal.pattern} ({signal.signal_type})',
                xaxis_rangeslider_visible=False,
                height=800
            )

            # Save chart
            chart_filename = f'charts/{symbol}_{datetime.now().strftime("%Y%m%d_%H%M%S")}.html'
            fig.write_html(chart_filename)

            return chart_filename

        except Exception as e:
            logging.error(f"Error creating visualization: {e}")
            return None

    def send_advanced_alert(self, analysis_result):
        """Send comprehensive alert with all information"""
        try:
            signal = analysis_result['signal']
            patterns = analysis_result['patterns']
            position_info = analysis_result['position_info']
            technical_data = analysis_result['technical_data']

            # Create comprehensive message
            message = f"""
üö® ADVANCED TRADING ALERT üö®
{'=' * 50}

üìä STOCK: {signal.symbol}
üéØ PATTERN: {signal.pattern}
üìà SIGNAL: {signal.signal_type}
üí™ CONFIDENCE: {signal.confidence:.1f}%
üß† MARKET REGIME: {signal.market_regime}
üòä SENTIMENT: {signal.sentiment_score:.2f} ({self.get_sentiment_label(signal.sentiment_score)})

üí∞ PRICE INFORMATION:
‚îú‚îÄ‚îÄ Current Price: ‚Çπ{signal.entry_price:.2f}
‚îú‚îÄ‚îÄ Target Price: ‚Çπ{signal.target_price:.2f} ({((signal.target_price/signal.entry_price-1)*100):.1f}%)
‚îú‚îÄ‚îÄ Stop Loss: ‚Çπ{signal.stop_loss:.2f} ({((signal.stop_loss/signal.entry_price-1)*100):.1f}%)
‚îî‚îÄ‚îÄ Risk:Reward = 1:{signal.risk_reward_ratio:.2f}

üìä TECHNICAL INDICATORS:
‚îú‚îÄ‚îÄ RSI: {technical_data.get('rsi', 'N/A'):.1f}
‚îú‚îÄ‚îÄ MACD: {technical_data.get('macd', 'N/A'):.4f}
‚îú‚îÄ‚îÄ Volume Ratio: {technical_data.get('volume_ratio', 'N/A'):.1f}x
‚îî‚îÄ‚îÄ ATR: {technical_data.get('atr', 'N/A'):.2f}

üíº POSITION SIZING:
‚îú‚îÄ‚îÄ Recommended Size: ‚Çπ{position_info['position_size']:,.0f}
‚îú‚îÄ‚îÄ Kelly Fraction: {position_info['kelly_fraction']:.2f}%
‚îú‚îÄ‚îÄ Risk Amount: ‚Çπ{position_info['risk_amount']:,.0f}
‚îî‚îÄ‚îÄ R:R Ratio: {position_info['risk_reward_ratio']:.2f}

üìã DETECTED PATTERNS:
{self.format_patterns_list(patterns)}

‚ö†Ô∏è CONFIRMATIONS:
‚îú‚îÄ‚îÄ Volume: {'‚úÖ' if signal.volume_confirmation else '‚ùå'}
‚îú‚îÄ‚îÄ Sentiment: {'‚úÖ' if signal.sentiment_score > 0 else '‚ùå'}
‚îî‚îÄ‚îÄ Market Regime: {signal.market_regime}

‚è∞ TIMESTAMP: {signal.timestamp.strftime('%Y-%m-%d %H:%M:%S')}

{'=' * 50}
‚ö†Ô∏è This is for educational purposes only. Trade responsibly!
            """

            # Send email alert
            self.send_email_alert_advanced(message, signal.symbol)

            # Log the alert
            logging.info(f"Advanced alert sent for {signal.symbol}")

            return message

        except Exception as e:
            logging.error(f"Error sending advanced alert: {e}")
            return None

    def format_patterns_list(self, patterns):
        """Format patterns list for display"""
        if not patterns:
            return "No patterns detected"

        formatted = ""
        for i, pattern in enumerate(patterns, 1):
            formatted += f"‚îú‚îÄ‚îÄ {i}. {pattern['pattern']} ({pattern.get('confidence', 0):.0f}%)\n"

        return formatted.rstrip('\n')

    def send_email_alert_advanced(self, message, symbol):
        """Send advanced email alert"""
        try:
            if not self.config['email']['sender_email']:
                return

            msg = MimeMultipart()
            msg['From'] = self.config['email']['sender_email']
            msg['To'] = ', '.join(self.config['email']['receiver_emails'])
            msg['Subject'] = f"üö® Advanced Stock Alert: {symbol}"

            msg.attach(MimeText(message, 'plain'))

            server = smtplib.SMTP(self.config['email']['smtp_server'], self.config['email']['smtp_port'])
            server.starttls()
            server.login(self.config['email']['sender_email'], self.config['email']['sender_password'])

            for receiver in self.config['email']['receiver_emails']:
                server.sendmail(self.config['email']['sender_email'], receiver, msg.as_string())

            server.quit()
            logging.info(f"Email alert sent successfully for {symbol}")

        except Exception as e:
            logging.error(f"Failed to send email alert: {e}")

    def run_advanced_scanner(self):
        """Run the advanced scanner with all features"""
        try:
            logging.info("üöÄ Starting Advanced Indian Stock Pattern Scanner...")

            # Scan stocks in parallel
            results = self.scan_stocks_parallel()

            if not results:
                logging.info("No significant patterns detected in this scan.")
                return

            # Sort by confidence
            results.sort(key=lambda x: x['signal'].confidence, reverse=True)

            logging.info(f"üéØ Found {len(results)} high-quality signals!")

            # Process each result
            for result in results:
                try:
                    # Send alert
                    alert_message = self.send_advanced_alert(result)

                    # Create visualization
                    symbol = result['signal'].symbol + '.NS'
                    data = self.get_enhanced_stock_data(symbol)
                    if data is not None:
                        chart_file = self.create_advanced_visualization(
                            symbol, data, result['patterns'], result['signal']
                        )
                        if chart_file:
                            logging.info(f"Chart saved: {chart_file}")

                    # Print summary
                    print(f"\n{'='*60}")
                    print(alert_message)
                    print(f"{'='*60}")

                except Exception as e:
                    logging.error(f"Error processing result: {e}")

            # Save results to file
            self.save_scan_results(results)

        except Exception as e:
            logging.error(f"Error in advanced scanner: {e}")

    def save_scan_results(self, results):
        """Save scan results to JSON file"""
        try:
            scan_data = {
                'timestamp': datetime.now().isoformat(),
                'total_signals': len(results),
                'signals': []
            }

            for result in results:
                signal = result['signal']
                scan_data['signals'].append({
                    'symbol': signal.symbol,
                    'pattern': signal.pattern,
                    'signal_type': signal.signal_type,
                    'confidence': signal.confidence,
                    'entry_price': signal.entry_price,
                    'target_price': signal.target_price,
                    'stop_loss': signal.stop_loss,
                    'risk_reward_ratio': signal.risk_reward_ratio,
                    'sentiment_score': signal.sentiment_score,
                    'market_regime': signal.market_regime,
                    'position_size': result['position_info']['position_size']
                })

            with open(f'scan_results_{datetime.now().strftime("%Y%m%d_%H%M%S")}.json', 'w') as f:
                json.dump(scan_data, f, indent=2)

            logging.info("Scan results saved successfully")

        except Exception as e:
            logging.error(f"Error saving scan results: {e}")

    def run_continuous_advanced_scanner(self, interval_minutes=30):
        """Run continuous advanced scanning"""
        def scan_job():
            try:
                logging.info(f"‚è∞ Starting scheduled scan at {datetime.now()}")
                self.run_advanced_scanner()
            except Exception as e:
                logging.error(f"Error in scheduled scan: {e}")

        # Schedule scans
        schedule.every(interval_minutes).minutes.do(scan_job)

        logging.info(f"üîÑ Continuous scanner started (every {interval_minutes} minutes)")

        # Run initial scan
        scan_job()

        # Keep running
        while True:
            try:
                schedule.run_pending()
                time.sleep(60)  # Check every minute
            except KeyboardInterrupt:
                logging.info("üõë Stopping continuous scanner...")
                break
            except Exception as e:
                logging.error(f"Error in continuous scanner: {e}")
                time.sleep(300)  # Wait 5 minutes before retrying

# Main execution
if __name__ == "__main__":
    # Create necessary directories
    import os
    os.makedirs('charts', exist_ok=True)

    # Initialize the advanced detector
    detector = AdvancedStockPatternDetector()

    print("üáÆüá≥ ADVANCED INDIAN STOCK PATTERN DETECTION SYSTEM")
    print("=" * 60)
    print("üöÄ Features:")
    print("‚îú‚îÄ‚îÄ Advanced Pattern Recognition (Harmonic, Elliott Wave, etc.)")
    print("‚îú‚îÄ‚îÄ Machine Learning Integration")
    print("‚îú‚îÄ‚îÄ Real-time Market Sentiment Analysis")
    print("‚îú‚îÄ‚îÄ Dynamic Position Sizing (Kelly Criterion)")
    print("‚îú‚îÄ‚îÄ Comprehensive Risk Management")
    print("‚îú‚îÄ‚îÄ Multi-threaded Analysis")
    print("‚îú‚îÄ‚îÄ Advanced Chart Visualization")
    print("‚îî‚îÄ‚îÄ Automated Email Alerts")
    print("=" * 60)

    choice = input("""
üéØ Choose Your Action:
1. üîç Run Single Advanced Scan
2. üîÑ Start Continuous Monitoring
3. üìä Analyze Specific Stock
4. üìà View Historical Performance
5. ‚öôÔ∏è  Configure Settings

Enter choice (1-5): """).strip()

    if choice == '1':
        print("\nüîç Running comprehensive market scan...")
        detector.run_advanced_scanner()

    elif choice == '2':
        interval = input("‚è∞ Enter scan interval in minutes (default 30): ").strip()
        interval = int(interval) if interval.isdigit() else 30
        detector.run_continuous_advanced_scanner(interval)

    elif choice == '3':
        symbol = input("üìä Enter stock symbol (e.g., RELIANCE, TCS): ").strip().upper()
        result = detector.analyze_single_stock(f"{symbol}.NS")

        if result:
            alert_message = detector.send_advanced_alert(result)
            print(f"\n{alert_message}")
        else:
            print(f"‚ùå No significant patterns found for {symbol}")

    elif choice == '4':
        print("üìà Historical performance analysis coming soon...")

    elif choice == '5':
        print("‚öôÔ∏è Configuration management coming soon...")

    print("\n‚úÖ Thank you for using Advanced Indian Stock Pattern Detection System!")
    print("üôè Trade responsibly and manage your risk!")
```
